```python
class Application(ctk.CTk):
    def __init__(self):
        super().__init__()
        self.title("Casa della Salute")
        self.geometry("1090x550+150+90")
        
        # Carica l'immagine usando PIL e convertila in ImageTk.PhotoImage
        pil_image = Image.open("C:/Users/nicol/OneDrive/Desktop/UNI/3 ANNO/Ingegneria del software/Divertimento/cc2.png")  # Assicurati che il percorso sia corretto
        self.bg_image = ImageTk.PhotoImage(pil_image)  # referenza all'immagine

        self.init_main_page()

    def init_main_page(self):
        self.clear_frame()

        # Label titolo
        title_label = ctk.CTkLabel(self, text="Casa di cura",
                                   fg_color = "#59B7D8"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        # Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image,
                                        text="Chi deve accedere: \n\n\n\n",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)

        # Bottoni all'interno del frame
        btn_personale = ctk.CTkButton(background_label, text="Personale della casa",
                                       command=self.show_personale_page,
                                       width = 200,
                                       height= 90,
                                       font=("Long Island", 40),
                                       hover_color = "#59B7D8")
        btn_personale.place( x = 70, y = 240)

        btn_paziente = ctk.CTkButton(background_label, text="Paziente",
                                    command=self.show_paziente_options,
                                    width = 200,
                                    height= 90,
                                    font=("Long Island", 40),
                                    hover_color = "#23F094"
                                    )
        btn_paziente.place( x = 780, y = 240)

    def show_personale_page(self):
        self.clear_frame()

        title_label = ctk.CTkLabel(self, text="Personale della Casa",
                                   fg_color = "#59B7D8"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        btn_back = ctk.CTkButton(title_label, text="Indietro", command=self.init_main_page,
                                 font=("Long Island", 24),
                                 )
        btn_back.pack(side="right", pady = 10, padx = 10)

        # Crea un Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image,
                                        text="Chi deve accedere: \n\n\n\n",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)
        
        btn_medico = ctk.CTkButton(self, text="Medico", command=self.show_medico_login_page,
                                   width = 200,
                                    height= 90,
                                     font=("Long Island", 40),
                                     hover_color = "#59B7D8")
        btn_medico.place( x = 100, y = 360)

        btn_infermiere = ctk.CTkButton(self, text="Infermiere", command=self.show_infermiere_login_page,
                                       width = 200,
                                    height= 90,
                                     font=("Long Island", 40),
                                     hover_color = "#59B7D8")
        btn_infermiere.place( x = 450, y = 360)

        btn_segreteria = ctk.CTkButton(self, text="Segreteria", command=self.show_segreteria_login_page,
                                       width = 200,
                                    height= 90,
                                     font=("Long Island", 40),
                                     hover_color = "#59B7D8")
        btn_segreteria.place( x = 800, y = 360)     
    
    def show_medico_login_page(self):
        self.clear_frame()

        title_label = ctk.CTkLabel(self, text="Medico",
                                   fg_color = "#59B7D8"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        btn_back = ctk.CTkButton(title_label, text="Indietro", command=self.show_personale_page,
                                 font=("Long Island", 24),
                                 )
        btn_back.pack(side="right", pady = 10, padx = 10)

        # Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image,
                                        text=" ",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)
        
        ctk.CTkLabel(background_label, text="Email",
                     font=("Long Island", 40),
                     ).place( x = 100, y = 100)
        self.entry_email = ctk.CTkEntry(background_label)
        self.entry_email.place(x = 100, y = 150)
        self.entry_email.focus_set()
        
        ctk.CTkLabel(background_label, text="Password",
                     font=("Long Island", 40)
                     ).place( x = 600, y = 100)
        self.entry_password = ctk.CTkEntry(background_label, show="*")
        self.entry_password.place(x = 600, y = 150)
        
        btn_login = ctk.CTkButton(background_label, text="Login", command=self.medico_logged_in,
                                  font=("Long Island", 40))
        btn_login.place(x = 100, y = 250)

        # Associa il tasto "Invio" alla funzione medico_logged_in
        self.bind('<Return>', self.medico_logged_in)

    def show_infermiere_login_page(self):
        self.clear_frame()

        title_label = ctk.CTkLabel(self, text="Infermiere",
                                   fg_color = "#59B7D8"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        btn_back = ctk.CTkButton(title_label, text="Indietro", command=self.show_personale_page,
                                 font=("Long Island", 24),
                                 )
        btn_back.pack(side="right", pady = 10, padx = 10)

        # Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image,
                                        text=" ",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)
        
        ctk.CTkLabel(background_label, text="Email",
                     font=("Long Island", 40),
                     ).place( x = 100, y = 100)
        self.entry_email = ctk.CTkEntry(background_label)
        self.entry_email.place(x = 100, y = 150)
        self.entry_email.focus_set()
        
        ctk.CTkLabel(background_label, text="Password",
                     font=("Long Island", 40)
                     ).place( x = 600, y = 100)
        self.entry_password = ctk.CTkEntry(background_label, show="*")
        self.entry_password.place(x = 600, y = 150)
        
        btn_login = ctk.CTkButton(background_label, text="Login", command=self.infermiere_logged_in,
                                  font=("Long Island", 40))
        btn_login.place(x = 100, y = 250)


        # Associa il tasto "Invio" alla funzione 
        self.bind('<Return>', self.infermiere_logged_in)

    def show_segreteria_login_page(self):
        self.clear_frame()

        title_label = ctk.CTkLabel(self, text="Segreteria",
                                   fg_color = "#59B7D8"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        btn_back = ctk.CTkButton(title_label, text="Indietro", command=self.show_personale_page,
                                 font=("Long Island", 24),
                                 )
        btn_back.pack(side="right", pady = 10, padx = 10)

        # Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image,
                                        text=" ",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)
        
        ctk.CTkLabel(background_label, text="Email",
                     font=("Long Island", 40),
                     ).place( x = 100, y = 100)
        self.entry_email = ctk.CTkEntry(background_label)
        self.entry_email.place(x = 100, y = 150)
        self.entry_email.focus_set()
        
        ctk.CTkLabel(background_label, text="Password",
                     font=("Long Island", 40)
                     ).place( x = 600, y = 100)
        self.entry_password = ctk.CTkEntry(background_label, show="*")
        self.entry_password.place(x = 600, y = 150)
        
        btn_login = ctk.CTkButton(background_label, text="Login", command=self.segreteria_logged_in,
                                  font=("Long Island", 40))
        btn_login.place(x = 100, y = 250)

        # Associa il tasto "Invio" alla funzione 
        self.bind('<Return>', self.segreteria_logged_in)

    def medico_logged_in(self, event=None):
        # Ottenere email e password inserite dall'utente
        email = self.entry_email.get()
        password = self.entry_password.get()

        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)
        query = "SELECT * FROM medici WHERE email = %s AND password = %s"
        cursor.execute(query, (email, password))

        # Ottenere il risultato della query
        medico = cursor.fetchone()

        if medico:
            MedicoInterface(self,medico,email,password)
            self.state('iconic')
        else:
            # Se le credenziali non esistono nel database, visualizza un messaggio di errore
            messagebox.showerror("Errore", "Accesso non riuscito. Controlla le credenziali e riprova.", parent = self)

        # Chiudere la connessione al database
        conn.close()

    def infermiere_logged_in(self, event=None):
        # Ottenere email e password inserite dall'utente
        email = self.entry_email.get()
        password = self.entry_password.get()

        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)
        query = "SELECT * FROM infermieri WHERE email = %s AND password = %s"
        cursor.execute(query, (email, password))

        # Ottenere il risultato della query
        infermiere = cursor.fetchone()

        if infermiere:
            InfermiereInterface(self,infermiere,email,password)
            self.state('iconic')
        else:
            # Se le credenziali non esistono nel database, visualizza un messaggio di errore
            messagebox.showerror("Errore", "Accesso non riuscito. Controlla le credenziali e riprova.", parent = self)

        # Chiudere la connessione al database
        conn.close()

    def segreteria_logged_in(self, event=None):
        # Ottenere email e password inserite dall'utente
        email = self.entry_email.get()
        password = self.entry_password.get()

        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)
        query = "SELECT * FROM segreteria WHERE email = %s AND password = %s"
        cursor.execute(query, (email, password))

        # Ottenere il risultato della query
        segreteria = cursor.fetchone()

        if segreteria:
            SegreteriaInterface(self,segreteria,email,password)
            self.state('iconic')
        else:
            # Se le credenziali non esistono nel database, visualizza un messaggio di errore
            messagebox.showerror("Errore", "Accesso non riuscito. Controlla le credenziali e riprova.", parent = self)

        # Chiudere la connessione al database
        conn.close()

    def show_paziente_options(self):
        self.clear_frame()

        title_label = ctk.CTkLabel(self, text="Pazienti",
                                   fg_color = "#48A259"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        btn_back =ctk.CTkButton(title_label, text="Indietro", command=self.init_main_page,
                                font=("Long Island", 24))
        btn_back.pack(side="right", pady = 10, padx = 10)

        # Carica l'immagine usando PIL 
        pil_image1 = Image.open("C:/Users/nicol/OneDrive/Desktop/UNI/3 ANNO/Ingegneria del software/Divertimento/gg.png")  # Assicurati che il percorso sia corretto
        self.bg_image1 = ImageTk.PhotoImage(pil_image1)  # Mantieni una referenza all'immagine

        # Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image1,
                                        text="Chi deve accedere: \n\n\n\n",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)

        btn_nuovo_paziente = ctk.CTkButton(background_label, text="Nuovo Paziente", command=self.show_registrazione_paziente,
                                            width = 200,
                                            height= 90,
                                            font=("Long Island", 40),
                                            fg_color= "#48A259",
                                            hover_color = "#95C799")
        btn_nuovo_paziente.place( x = 70, y =240)

        btn_accedi_paziente = ctk.CTkButton(background_label, text="Accedi", command=self.show_paziente_login_page,
                                             width = 200,
                                            height= 90,
                                            font=("Long Island", 40),
                                            fg_color= "#48A259",
                                            hover_color = "#23F094")
        btn_accedi_paziente.place( x = 780, y =240)
        
    def show_registrazione_paziente(self):
        NuovoPaziente(self)
        self.state('iconic')
    
    def show_paziente_login_page(self, event = None):
        self.clear_frame()

        title_label = ctk.CTkLabel(self, text="Pazienti",
                                   fg_color = "#48A259"
                                   ,font=("Long Island", 50),
                                   text_color="Black",
                                   height = 100)
        title_label.pack(fill="both")
        title_label.pack_propagate(False)

        btn_back = ctk.CTkButton(title_label, text="Indietro", command=self.show_paziente_options,
                                 fg_color = "#48A259",
                                 font=("Long Island", 24),
                                 )
        btn_back.pack(side="right", pady = 10, padx = 10)

        # Label per usare l'immagine come sfondo
        background_label = ctk.CTkLabel(self, image=self.bg_image1,
                                        text=" ",
                                        font=("Long Island", 40))
        background_label.place(x=0, y=100, relwidth=1, relheight=1)
        
        ctk.CTkLabel(background_label, text="Email",
                     font=("Long Island", 40),
                     ).place( x = 100, y = 100)
        self.entry_email = ctk.CTkEntry(background_label)
        self.entry_email.place(x = 100, y = 150)
        self.entry_email.focus_set()
        
        ctk.CTkLabel(background_label, text="Password",
                     font=("Long Island", 40)
                     ).place( x = 600, y = 100)
        self.entry_password = ctk.CTkEntry(background_label, show="*")
        self.entry_password.place(x = 600, y = 150)
        
        btn_login = ctk.CTkButton(background_label, text="Login", command=self.paziente_logged_in,
                                  font=("Long Island", 40))
        btn_login.place(x = 100, y = 250)

        # Associa il tasto "Invio" alla funzione 
        self.bind('<Return>', self.paziente_logged_in)

    def paziente_logged_in(self, event = None):
        # Ottenere email e password inserite dall'utente
        email = self.entry_email.get()
        password = self.entry_password.get()

        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)
        query = "SELECT * FROM pazienti WHERE email = %s AND password = %s"
        cursor.execute(query, (email, password))

        # Ottenere il risultato della query
        paziente = cursor.fetchone()

        if paziente:
            PazienteInterface(self,paziente,email,password)
            self.state('iconic')
        else:
            # Se le credenziali non esistono nel database, visualizza un messaggio di errore
            messagebox.showerror("Errore", "Accesso non riuscito. Controlla le credenziali e riprova.", parent = self)

        # Chiudere la connessione al database
        conn.close()
    
    def clear_frame(self):
        for widget in self.winfo_children():
            widget.destroy()

class ModificaDatiDialog(ctk.CTkToplevel):
    def __init__(self, parent, tipo, medico_id=None, paziente_id=None, is_paziente=False):
        super().__init__(parent)
        self.tipo = tipo
        self.medico_id = medico_id
        self.paziente_id = paziente_id
        self.is_paziente = is_paziente  # Determina se il chiamante è un paziente
        self.title(f"Modifica {tipo.capitalize()}")
        self.geometry("400x300")

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        self.id_selezionato = paziente_id

        ctk.CTkLabel(self, text=f"Seleziona un {tipo} dalla lista:").pack(pady=5)

        self.listbox_utenti = tk.Listbox(self, width=100)
        self.listbox_utenti.pack(pady=5)
        self.listbox_utenti.bind('<<ListboxSelect>>', self.seleziona_utente)

        self.button_modifica = ctk.CTkButton(self, text="Modifica", command=self.modifica_utente)
        self.button_modifica.pack(pady=20)

        self.ottieni_utenti()

    def ottieni_utenti(self):
        conn = connessione_db()
        c = conn.cursor()

        # Modifica la query in base a se il chiamante è un paziente o un medico
        if self.is_paziente and self.paziente_id:
            # Ottieni figli a carico se il chiamante è un paziente
            c.execute("SELECT p.ID, p.Nome, p.Cognome FROM pazienti p JOIN responsabilita r ON p.ID = r.minorenne_id WHERE r.maggiorenne_id = %s", (self.paziente_id,))
        elif self.tipo == "pazienti" and self.medico_id:
            # Ottieni pazienti del medico
            c.execute("SELECT ID, Nome, Cognome FROM pazienti WHERE medico_curante = %s", (self.medico_id,))
        else:
            c.execute(f"SELECT ID, Nome, Cognome FROM {self.tipo}")
        
        utenti = c.fetchall()
        conn.close()

        for utente in utenti:
            # Converte ogni elemento di 'utente' in stringa e li unisce con il carattere " - "
            # map(str, utente) trasforma tutti gli elementi della lista 'utente' in stringhe
            # " - ".join(...) unisce questi elementi con " - " come separatore
            utente_str = " - ".join(map(str, utente))
            # Inserisce la stringa 'utente_str' nella listbox 'listbox_utenti' alla fine
            # tk.END rappresenta la posizione finale della listbox
            self.listbox_utenti.insert(tk.END, utente_str)

    def seleziona_utente(self, event):
        index = self.listbox_utenti.curselection()
        if index:
            self.id_selezionato = self.listbox_utenti.get(index).split(" - ")[0]

    def modifica_utente(self):
        if not self.id_selezionato:
            messagebox.showerror("Errore", f"Seleziona un {self.tipo} dalla lista.", parent = self)
            return

        ModificaDettagliDialog(self, self.tipo, self.id_selezionato)

class ModificaDettagliDialog(ctk.CTkToplevel):
    def __init__(self, parent, tipo=None, id_selezionato=None):
        super().__init__(parent)
        self.tipo = tipo # self.tipo = pazienti
        self.id_selezionato = id_selezionato
        self.title(f"Modifica {tipo.capitalize()} Dettagli")
        self.geometry("450x500")

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        # Connessione al database tramite funzione esterna
        self.conn = connessione_db()
        self.c = self.conn.cursor()
        
        self.c.execute(f"SELECT * FROM {self.tipo} WHERE ID = %s", (self.id_selezionato,))

        dati = self.c.fetchone()

        # Frame principale
        self.frame = tk.Frame(self)
        self.frame.pack(fill="both", expand=True, padx=10, pady=10)

        self.labels_entries = []
        for i, col in enumerate(self.c.column_names):
            if col == 'ID' or col == '':
                continue

            # Crea etichetta e entry/checkbox per ogni colonna
            label = ctk.CTkLabel(self.frame, text=col)
            label.grid(row=i, column=0, padx=10, pady=5, sticky="w")  # Aggiungi la label alla griglia

            if col == 'responsabilita':
                var_responsabilita = tk.BooleanVar(value=bool(dati[i]))
                check_responsabilita = tk.Checkbutton(self.frame, variable=var_responsabilita, text="Responsabilità (Sì/No)")
                check_responsabilita.grid(row=i, column=1, padx=10, pady=5, sticky="w")
                self.labels_entries.append((col, var_responsabilita))
            else:
                entry = tk.Entry(self.frame)
                entry.grid(row=i, column=1, padx=10, pady=5, sticky="w")
                entry.insert(0, dati[i])
                self.labels_entries.append((col, entry))

        # Bottone Salva, posizionato nella griglia
        save_button = ctk.CTkButton(self.frame, text="Salva", command=self.salva_modifiche)
        save_button.grid(row=len(self.c.column_names), column=0, columnspan=2, pady=20)

    def salva_modifiche(self):

        # 'self.tipo' rappresenta il nome della tabella che verrà aggiornata
        update_query = f"UPDATE {self.tipo} SET "
        # Lista per memorizzare i valori da inserire nella query
        update_data = []

        # Itera attraverso le colonne e i relativi widget
        for col, widget in self.labels_entries:
            # Controlla se il widget è una variabile booleana
            if isinstance(widget, tk.BooleanVar):
                # Aggiunge alla query la parte per aggiornare questa colonna
                update_query += f"{col} = %s, "
                # Aggiunge il valore corrente del widget (True/False) alla lista dei dati
                update_data.append(widget.get())
            else:
                # Aggiunge alla query la parte per aggiornare la colonna
                update_query += f"{col} = %s, "
                # Aggiunge il valore corrente del widget alla lista dei dati
                update_data.append(widget.get())

        # Rimuove l'ultima virgola e spazio superflui dalla query
        update_query = update_query.rstrip(", ")
        # Aggiunge la condizione WHERE per specificare quale riga aggiornare (in base all'ID selezionato)
        update_query += f" WHERE ID = %s"
        # Aggiunge l'ID selezionato alla lista dei dati (necessario per l'istruzione WHERE)
        update_data.append(self.id_selezionato)

        try:
            self.c.execute(update_query, update_data)
            self.conn.commit()
            messagebox.showinfo("Successo", f"{self.tipo.capitalize()} aggiornato con successo!", parent = self)
            self.destroy()
        except mysql.connector.Error as err:
            messagebox.showerror("Errore", f"Errore durante l'aggiornamento: {err}", parent = self)
        finally:
            self.conn.close()

# medico
class MedicoInterface(ctk.CTkToplevel):
    def __init__(self, parent, medico,email, password): #medico è stato passato come dizionario
        super().__init__(parent)
        self.title("Interfaccia Medico")
        self.geometry("1200x600+70+50")

        self.medico_id = medico.get('id') 
        self.nome = medico.get('nome')
        self.cognome = medico.get('cognome')
        self.medico = medico
        self.email = email
        self.password = password

        # Connessione al database
        self.conn = connessione_db()

        self.create_widgets()

    def create_widgets(self):

        self.clear_frame()

        c = self.conn.cursor()
        c.execute("SELECT nome, cognome FROM medici WHERE email = %s AND password = %s", (self.email, self.password))
        medico_data = c.fetchone()
        c.close()

        nome, cognome = medico_data

        # Frame superiore per nome e cognome del medico
        frame_top = ctk.CTkFrame(self, fg_color="#59B7D8", height=50)
        frame_top.pack(fill="x")

        # Nome e cognome del medico
        nome_label = ctk.CTkLabel(frame_top, text=f"Dr. {nome} {cognome}", font=("Long Island", 20))
        nome_label.grid(row=0, column=0, padx=(10, 5), pady=10, sticky="w")

        # Creazione del bottone per il logout
        logout_button = ctk.CTkButton(frame_top, text="Logout", command=self.logout)
        logout_button.grid(row=0, column=1, padx=10, pady=10, sticky="e")

        # Allinea tutto in una griglia con lo spazio rimanente
        frame_top.grid_columnconfigure(0, weight=1)

        # Frame inferiore per i bottoni
        frame_bottom = ctk.CTkFrame(self, height = 50, fg_color="#59B7D8")
        frame_bottom.pack(side="bottom", fill="x")

        # Frame sinistro per il notebook principale
        frame_left = ctk.CTkFrame(self, width=450, fg_color="white")
        frame_left.pack(side="left", fill="both", expand=True)
        frame_left.pack_propagate(False)

        # Creazione del Tabview principale
        notebook_main = ctk.CTkTabview(frame_left)
        notebook_main.pack(fill="both", expand=True)

        # Tab per "Pazienti"
        tab_pazienti = notebook_main.add("Pazienti")
        sub_notebook_pazienti = ctk.CTkTabview(tab_pazienti)
        sub_notebook_pazienti.pack(fill="both", expand=True)

        sub_tab_miei_pazienti = sub_notebook_pazienti.add("Miei")

        # Tab "Pazienti miei"
        frame_bottoni_pazienti = ctk.CTkFrame(sub_tab_miei_pazienti)
        frame_bottoni_pazienti.pack(side="bottom", pady=10)
        ctk.CTkButton(frame_bottoni_pazienti, text="Modifica Paziente", command=lambda: self.apri_modifica_dialog("pazienti"), fg_color="orange", text_color="black").grid(row=0, column=1, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_pazienti, text="Aggiorna", command=self.populate_miei_pazienti, fg_color="orange", text_color="black").grid(row=0, column=2, padx=10, pady=10)
        # Crea e configura la Listbox
        self.listbox_pazienti_miei = tk.Listbox(sub_tab_miei_pazienti, height= 50)
        self.listbox_pazienti_miei.pack(side='top', fill='x')
        self.populate_miei_pazienti()

        sub_tab_tutti_pazienti = sub_notebook_pazienti.add("Tutti")

        # Frame bottoni "Pazienti"
        frame_bottoni_pazienti = ctk.CTkFrame(sub_tab_tutti_pazienti)
        frame_bottoni_pazienti.pack(side="bottom", pady=10)
        ctk.CTkButton(frame_bottoni_pazienti, text="Modifica Paziente", command=lambda: self.apri_modifica_dialog("pazienti"), fg_color="orange", text_color="black").grid(row=0, column=1, padx=10, pady=10)
        
        # Crea e configura la Listbox
        self.listbox_pazienti_tutti = tk.Listbox(sub_tab_tutti_pazienti, height= 50)
        self.listbox_pazienti_tutti.pack(side='top', fill='x')
        self.populate_tutti_pazienti()
        ctk.CTkButton(frame_bottoni_pazienti, text="Aggiorna", command=self.populate_tutti_pazienti, fg_color="orange", text_color="black").grid(row=0, column=2, padx=10, pady=10)

        # Tab per "Assenze"
        tab_assenze = notebook_main.add("Assenze")
        sub_notebook_assenze = ctk.CTkTabview(tab_assenze)
        sub_notebook_assenze.pack(fill="both", expand=True)

        sub_tab_miei_assenze = sub_notebook_assenze.add("Mie assenze")
        self.textbox_miei_assenze = ctk.CTkTextbox(sub_tab_miei_assenze, width=200, wrap="word")
        self.textbox_miei_assenze.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_miei_assenze.configure(state="disabled") # Rende il textbox non modificabile dall'utente
        self.populate_miei_assenze()

        sub_tab_tutti_assenze = sub_notebook_assenze.add("Tutti")
        self.textbox_tutti_assenze = ctk.CTkTextbox(sub_tab_tutti_assenze, width=200, wrap="word")
        self.textbox_tutti_assenze.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutti_assenze.configure(state="disabled") # Rende il textbox non modificabile dall'utente
        self.populate_tutti_assenze()

        # Tab per "Visite, Prelievi e Medicazioni"
        tab_visite = notebook_main.add("Visite, Prelievi e Medicazioni")
        sub_notebook_visite = ctk.CTkTabview(tab_visite)
        sub_notebook_visite.pack(fill="both", expand=True)

        sub_tab_miei_esiti = sub_notebook_visite.add("Mie visite")
        self.textbox_miei_esiti = ctk.CTkTextbox(sub_tab_miei_esiti, width=200, wrap="word")
        self.textbox_miei_esiti.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_miei_esiti.configure(state="disabled") # Rende il textbox non modificabile dall'utente
        self.populate_miei_esiti()

        sub_tab_tutti_esiti = sub_notebook_visite.add("Tutte le visite")
        self.textbox_tutti_esiti = ctk.CTkTextbox(sub_tab_tutti_esiti, width=200, wrap="word")
        self.textbox_tutti_esiti.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutti_esiti.configure(state="disabled") # Rende il textbox non modificabile dall'utente
        self.populate_esiti()

        # Frame centrale per i dettagli del paziente
        self.frame_center = ctk.CTkFrame(self, width=300, fg_color="white")
        self.frame_center.pack(side="left", fill="both", expand=True)
        self.frame_center.pack_propagate(False)
        
        # Frame destro per "Prenotazioni"
        frame_right = ctk.CTkFrame(self, width=350, fg_color="white")
        frame_right.pack(side="right", fill="both", expand=True)
        frame_right.pack_propagate(False)
        ctk.CTkLabel(frame_right, text="Prenotazioni", font=("Long Island", 16)).pack(pady=10)
        
        # Creazione del Tabview per le prenotazioni
        notebook_comunicazioni = ctk.CTkTabview(frame_right)
        notebook_comunicazioni.pack(fill="both", expand=True)

        # Tab per "Storico"
        tab_mie_comunicazioni = notebook_comunicazioni.add("Storico")
        self.textbox_mie_comunicazioni = ctk.CTkTextbox(tab_mie_comunicazioni, width=200, height=350, wrap="word")
        self.textbox_mie_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_mie_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        # Tab per "Prossime prenotazioni"
        tab_tutte_comunicazioni = notebook_comunicazioni.add("Prossime prenotazioni")
        self.textbox_tutte_comunicazioni = ctk.CTkTextbox(tab_tutte_comunicazioni, width=200, height=350, wrap="word")
        self.textbox_tutte_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutte_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        ctk.CTkButton(frame_right, text="Aggiorna", command=self.aggiorna_comunicazioni_medico).pack()

        ctk.CTkButton(frame_bottom, text="Inserisci Esito", command=self.apri_finestra_inserisci_esito).pack(side="bottom", padx=10, pady=10)
        
        # Creazione di un frame per i bottoni nella tab 'Assenze'
        frame_bottoni_assenze = ctk.CTkFrame(tab_assenze, height = 200)
        frame_bottoni_assenze.pack(side="bottom", pady=10)
        frame_bottoni_assenze.pack_propagate(False)

        # Inserimento di bottoni in una griglia 2x2
        ctk.CTkButton(frame_bottoni_assenze, text="Inserisci Assenza", command=self.open_assenza_dialog).grid(row=0, column=0, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_assenze, text="Elimina Assenza", command=self.elimina_assenza).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_assenze, text="Aggiorna", command=self.aggiorna_assenze).grid(row=1, column=0, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_assenze, text="Modifica assenza", command=self.modifica_assenza).grid(row=1, column=1, padx=10, pady=10, sticky="ew")

        # Configurazione della griglia per distribuire uniformemente lo spazio
        frame_bottoni_assenze.grid_columnconfigure(0, weight=1)
        frame_bottoni_assenze.grid_columnconfigure(1, weight=1)

        # Per assicurarsi che i bottoni riempiano lo spazio disponibile, è possibile configurare il frame per espandere
        frame_bottoni_assenze.pack(fill="x")

        # Creazione di un frame per i bottoni nella tab 'visite'
        frame_bottoni_visite = ctk.CTkFrame(tab_visite, height = 200)
        frame_bottoni_visite.pack(side="bottom", pady=10)
        frame_bottoni_visite.pack_propagate(False)

        # Inserimento di bottoni in una griglia 2x2
        ctk.CTkButton(frame_bottoni_visite, text="Elimina Esito", command=lambda: self.elimina_esito(self.medico_id)).grid(row=0, column=0, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_visite, text="Aggiorna", command=self.aggiorna_esito).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_visite, text="Modifica Esito", command=lambda:self.modifica_esito(self.medico_id)).grid(row=1, column=0, padx=10, pady=10, sticky="ew")

        # Configurazione della griglia per distribuire uniformemente lo spazio
        frame_bottoni_visite.grid_columnconfigure(0, weight=1)
        frame_bottoni_visite.grid_columnconfigure(1, weight=1)

        # Per assicurarsi che i bottoni riempiano lo spazio disponibile, è possibile configurare il frame per espandere
        frame_bottoni_visite.pack(fill="x")

        # Aggiornare le comunicazioni all'avvio
        self.aggiorna_comunicazioni_medico()

    def apri_finestra_inserisci_esito(self):
        InserisciEsitoDialog(self, self.medico_id)

    def apri_modifica_dialog(self, tipo):
        ModificaDatiDialog(self, tipo, self.medico_id)

    def open_assenza_dialog(self):
        AggiungiAssenzaDialog(self, self.medico_id)

    def elimina_assenza(self):
        EliminaAssenzaDialog(self, self.medico_id)
    
    def aggiorna_comunicazioni_medico(self):
        oggi = datetime.today().date()
        c = self.conn.cursor()

        # Recuperare le prenotazioni del medico che sono antecedenti a oggi
        c.execute("""
            SELECT p.data, p.orario, p.sala, p.tipo_visita, pa.Nome, pa.Cognome
            FROM prenotazioni p
            JOIN medici m ON p.medico_id = m.id
            JOIN pazienti pa ON p.paziente_id = pa.id
            WHERE p.medico_id = %s AND p.data < %s
            ORDER BY p.data DESC, p.orario
        """, (self.medico_id, oggi))
        mie_prenotazioni_passate = c.fetchall()

        # Recuperare le prenotazioni future del medico
        c.execute("""
            SELECT p.data, p.orario, p.sala, p.tipo_visita, pa.Nome, pa.Cognome, m.Nome, m.Cognome
            FROM prenotazioni p
            JOIN medici m ON p.medico_id = m.id
            JOIN pazienti pa ON p.paziente_id = pa.id
            WHERE p.medico_id = %s AND p.data >= %s
            ORDER BY p.data, p.orario
        """, (self.medico_id, oggi))
        tutte_prenotazioni_future = c.fetchall()
        
        c.close()

        # Aggiornare il contenuto delle etichette con le prenotazioni
        mie_prenotazioni_passate_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')} alle {orario} in {sala} visita {tipo_visita} \ncon {nome} {cognome}" for data, orario, sala, tipo_visita, nome, cognome in mie_prenotazioni_passate]
        ) if mie_prenotazioni_passate else "Nessuna prenotazione registrata."

        tutte_prenotazioni_future_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')} alle {orario} in {sala} visita {tipo_visita} \ncon {paz_nome} {paz_cognome}" for data, orario, sala, tipo_visita, paz_nome, paz_cognome, med_nome, med_cognome in tutte_prenotazioni_future]
        ) if tutte_prenotazioni_future else "Nessuna prenotazione."

        # Aggiornare il contenuto delle Textbox

        # Imposta la textbox in stato "normale" (abilitata), permettendo modifiche al suo contenuto
        self.textbox_mie_comunicazioni.configure(state="normal")
        # Cancella il contenuto esistente nella textbox. 
        # "1.0" rappresenta l'inizio del testo (riga 1, carattere 0) e "end" indica la fine del testo
        self.textbox_mie_comunicazioni.delete("1.0", "end")
        # Inserisce il nuovo contenuto nella textbox, partendo dall'inizio ("1.0").
        # 'mie_prenotazioni_passate_str' è la stringa che contiene il testo da visualizzare
        self.textbox_mie_comunicazioni.insert("1.0", mie_prenotazioni_passate_str)
        # Reimposta la textbox in stato "disabilitato" (non modificabile), impedendo ulteriori modifiche da parte dell'utente
        self.textbox_mie_comunicazioni.configure(state="disabled")

        self.textbox_tutte_comunicazioni.configure(state="normal")
        self.textbox_tutte_comunicazioni.delete("1.0", "end")
        self.textbox_tutte_comunicazioni.insert("1.0", tutte_prenotazioni_future_str)
        self.textbox_tutte_comunicazioni.configure(state="disabled")

    def populate_miei_esiti(self):
        
        # Pulisci la Textbox prima di inserire i nuovi dati
        self.textbox_miei_esiti.configure(state="normal")
        self.textbox_miei_esiti.delete("1.0", "end")

        conn = connessione_db()
        c = conn.cursor()
        c.execute("""
        SELECT esito, data, ambulatorio, urgenza, tipo_visita, p.nome, p.cognome
        FROM esiti
        LEFT JOIN pazienti p ON esiti.paziente_id = p.id
        WHERE medico_id = %s
        ORDER BY data DESC
        """, (self.medico_id,))
        esiti = c.fetchall()
        c.close()

        # Inserisci i nuovi dati
        if esiti:
            for esito, data, ambulatorio, urgenza, tipo_visita, nome, cognome in esiti:
                self.textbox_miei_esiti.insert("end", f"- {data} \n{ambulatorio} \nUrgenza: {urgenza} \nTipo visita: {tipo_visita}\nPaziente: {nome} {cognome}\nEsito: {esito}\n\n")
        else:
            self.textbox_miei_esiti.insert("end", "Nessuna assenza registrata.")

        # Imposta lo stato di 'disabled' per evitare modifiche
        #self.textbox_miei_esiti.configure(state="disabled")

    def populate_esiti(self):
        # Pulisci la Textbox prima di inserire i nuovi dati
        self.textbox_tutti_esiti.configure(state="normal")
        self.textbox_tutti_esiti.delete("1.0", "end")

        conn = connessione_db()
        c = conn.cursor()
        c.execute("""
            SELECT e.esito, e.data, e.ambulatorio, e.urgenza, e.tipo_visita, 
                CONCAT(p.nome, ' ', p.cognome) AS nome_completo_paziente,
                COALESCE(m.Nome, i.Nome, s.Nome) AS Nome,
                COALESCE(m.Cognome, i.Cognome, s.Cognome) AS Cognome
            FROM esiti e
            LEFT JOIN medici m ON e.medico_id = m.id
            LEFT JOIN infermieri i ON e.infermiere_id = i.id
            LEFT JOIN segreteria s ON e.segreteria_id = s.id
            LEFT JOIN pazienti p ON e.paziente_id = p.id
            ORDER BY e.data DESC
        """)
        # Si utilizza COALESCE per ottenere il primo valore non nullo tra i campi nome e cognome delle tre tabelle

        esiti = c.fetchall()

        c.close()

        # Inserisci i nuovi dati
        if esiti:
            for esito, data, ambulatorio, urgenza, tipo_visita, nome, cognome, nome_completo_paziente in esiti:
                self.textbox_tutti_esiti.insert("end", f"- {data} \n{ambulatorio} \nUrgenza: {urgenza} \nTipo visita: {tipo_visita} \nCaricato da: {cognome}  {nome_completo_paziente}\nPaziente: {nome}\nEsito: {esito}\n\n")
        else:
            self.textbox_tutti_esiti.insert("end", "Nessun esito registrato.")

        # Imposta lo stato di 'disabled' per evitare modifiche
        self.textbox_tutti_esiti.configure(state="disabled")

    def populate_miei_pazienti(self):

        # Pulisci la Textbox prima di inserire i nuovi dati
        #self.listbox_pazienti_miei.configure(state="normal")
        self.listbox_pazienti_miei.delete(0,  tk.END)

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM pazienti WHERE medico_curante = %s ORDER BY cognome", (self.medico_id,))
        pazienti = c.fetchall()

        for paziente in pazienti:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_pazienti_miei.insert(tk.END, f"{paziente[0]} {paziente[2]} {paziente[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_pazienti_miei.bind('<<ListboxSelect>>', self.seleziona_utente)

        # Imposta lo stato di 'disabled' per evitare modifiche
        #self.listbox_pazienti_miei.configure(state="disabled")
        c.close()

    def seleziona_utente(self, event):
        # Ottieni l'indice dell'elemento selezionato nella listbox 'listbox_pazienti_miei'
        index = self.listbox_pazienti_miei.curselection()
        if index:
            # Ottieni l'elemento selezionato (in base all'indice) e dividilo in base al carattere " - ".
            # 'split(" - ")' suddivide la stringa e prende il primo elemento (che rappresenta l'ID del paziente)
            self.id_selezionato = self.listbox_pazienti_miei.get(index[0]).split(" - ")[0]
            # Richiama la funzione 'show_patient_details', passando l'ID selezionato per visualizzare i dettagli del paziente
            self.show_patient_details(self.id_selezionato)

    def populate_tutti_pazienti(self):

        self.listbox_pazienti_tutti.delete(0,  tk.END)

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM pazienti ORDER BY cognome")
        pazienti = c.fetchall()

        for paziente in pazienti:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_pazienti_tutti.insert(tk.END, f"{paziente[0]} {paziente[2]} {paziente[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_pazienti_tutti.bind('<<ListboxSelect>>', self.seleziona_utente2)

        c.close()

    def seleziona_utente2(self, event):
        # Ottieni l'indice dell'elemento selezionato nella listbox 
        index = self.listbox_pazienti_tutti.curselection()
        if index:
            # Ottieni l'elemento selezionato (in base all'indice) e dividilo in base al carattere " - ".
            # 'split(" - ")' suddivide la stringa e prende il primo elemento (che rappresenta l'ID del paziente)
            self.id_selezionato = self.listbox_pazienti_tutti.get(index[0]).split(" - ")[0]
            # Richiama la funzione 'show_patient_details', passando l'ID selezionato per visualizzare i dettagli del paziente
            self.show_patient_details(self.id_selezionato)

    def show_patient_details(self, patient_id):
        for widget in self.frame_center.winfo_children():
            widget.destroy()

        conn = connessione_db()
        
        c = conn.cursor()
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico ,r.minorenne_id
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id 
                LEFT JOIN responsabilita r ON p.id = r.maggiorenne_id
                WHERE p.id = %s
            """, (patient_id,))
        paziente = c.fetchone()
        conn.close()

        if paziente:
            ctk.CTkLabel(self.frame_center, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Password: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(self.frame_center, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=10)


            # Controlla se il paziente ha minorenni a carico
            if paziente[11] is not None:
                ctk.CTkButton(self.frame_center, text="Visualizza Responsabilità", command=lambda: self.visualizza_responsabilita(paziente[0])).pack(pady=10)
                
        else:
            ctk.CTkLabel(self.frame_center, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)
        c.close()

    def show_history(self, id = None):
        self.clear_frame()

        # Textbox
        self.text_box = ctk.CTkTextbox(self, height=10, wrap="word", state="normal", yscrollcommand=True)
        self.text_box.pack(pady=20, fill="both", expand=True)

        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        conn = connessione_db()
        cursor = conn.cursor()

        # Se viene passato l'id paziente mostra gli esiti del paziente
        if id: 
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                WHERE p.id = %s
                ORDER BY e.data DESC
            """, (id,))
        else:
            # Altrimenti mostra tutti gli esiti di tutti i pazienti
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                ORDER BY e.data DESC
            """)
        rows = cursor.fetchall()
        conn.close()

        self.text_box.delete('1.0', tk.END)  # Pulisce il textbox prima di inserire nuovi dati
        if not rows:  # Controlla se la lista è vuota
            self.text_box.insert(tk.END, "Nessun esito registrato")  # Mostra messaggio se non ci sono dati
        else:
            for row in rows:
                urgenza_status = "No" if row[2] == 0 else "Yes"
                healthcare_provider = f"Medico: {row[5]} {row[6]}" if row[5] and row[6] else f"Infermiere: {row[7]} {row[8]}"
                formatted_row = (f"- {row[0]} \t Ambulatorio: {row[1]} \t Urgenza: {urgenza_status} \t "
                                f"Regime della visita: {row[3]} \n"
                                f"{healthcare_provider} \n"
                                f"Paziente: {row[9]} {row[10]}\nEsito: {row[4]} \n\n")
                self.text_box.insert(tk.END, formatted_row) 

    def visualizza_responsabilita(self, id):
        self.clear_frame()  # Pulisce i widget esistenti

        # Bottone per tornare alla vista principale
        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        frame_responsabilita = ctk.CTkFrame(self, width=350)
        frame_responsabilita.pack(side="left", fill="both", expand=True)
        frame_responsabilita.pack_propagate(False)

        # Connessione al database MySQL
        conn = connessione_db()
        c = conn.cursor() 

        # Query per recuperare i dettagli del paziente maggiorenne e il conteggio dei minorenni a carico
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico,
                    (SELECT COUNT(*) FROM responsabilita WHERE maggiorenne_id = p.id) AS count_minorenni
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id
                WHERE p.id = %s
            """, (id,))
        paziente = c.fetchone()

        # Query per ottenere i dettagli di ogni minorenne a carico
        c.execute("""
                SELECT p.*
                FROM pazienti p
                JOIN responsabilita r ON p.id = r.minorenne_id
                WHERE r.maggiorenne_id = %s
            """, (id,))
        minorenni = c.fetchall()

        # Mostra i dettagli del maggiorenne
        if paziente:
            ctk.CTkLabel(frame_responsabilita, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Password: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(frame_responsabilita, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=5)

            # Mostra i dettagli di ogni minorenne
            for minorenne in minorenni:
                min_frame = ctk.CTkFrame(self, width=350)
                min_frame.pack(side = "left",  fill="both", expand=True)
                min_frame.pack_propagate(False)
                ctk.CTkLabel(min_frame, text=f"Minorenne - Nome: {minorenne[1]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Cognome: {minorenne[2]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"CF: {minorenne[3]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Data di nascita: {minorenne[5]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Luogo di nascita: {minorenne[6]}").pack(pady=5)
                ctk.CTkButton(min_frame, text="Storico", command=lambda min_id=minorenne[0]: self.show_history(min_id)).pack(pady=5)

        else:
            ctk.CTkLabel(frame_responsabilita, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

        conn.close()

    def aggiorna_assenze(self):
        self.populate_miei_assenze()
        self.populate_tutti_assenze()

    def populate_miei_assenze(self):
        
        # Pulisci la Textbox prima di inserire i nuovi dati
        self.textbox_miei_assenze.configure(state="normal")
        self.textbox_miei_assenze.delete("1.0", "end")

        # Ristabilire la connessione e riscaricare i dati
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT inizio, fine FROM assenze WHERE medico = %s ORDER BY inizio ASC", (self.medico_id,))
        mie_assenze = c.fetchall()
        conn.close()  # Chiudi la connessione dopo l'uso

        # Inserisci i nuovi dati
        if mie_assenze:
            for inizio, fine in mie_assenze:
                self.textbox_miei_assenze.insert("end", f"- Dal {inizio} al {fine}\n")
        else:
            self.textbox_miei_assenze.insert("end", "Nessuna assenza registrata.")

        # Imposta lo stato di 'disabled' per evitare modifiche
        self.textbox_miei_assenze.configure(state="disabled")

    def populate_tutti_assenze(self):
        # Pulisci la Textbox prima di inserire i nuovi dati
        self.textbox_tutti_assenze.configure(state="normal")
        self.textbox_tutti_assenze.delete("1.0", "end")

        # Ristabilire la connessione e riscaricare i dati
        conn = connessione_db()
        c = conn.cursor()
        c.execute("""SELECT medici.nome, medici.cognome, assenze.inizio, assenze.fine 
                  FROM assenze 
                  JOIN medici ON assenze.medico = medici.id 
                  ORDER BY assenze.inizio ASC""")
        assenze = c.fetchall()
        
        conn.close()  # Chiudi la connessione dopo l'uso

        # Inserisci i nuovi dati
        if assenze:
            for nome, cognome, inizio, fine in assenze:
                self.textbox_tutti_assenze.insert("end", f"- {nome} {cognome}: Dal {inizio} al {fine}\n")
        else:
            self.textbox_tutti_assenze.insert("end", "Nessuna assenza registrata.")

        # Imposta lo stato di 'disabled' per evitare modifiche
        self.textbox_tutti_assenze.configure(state="disabled")

    def modifica_assenza(self):
        ModificaAssenza(self, self.medico_id)

    def elimina_esito(self, medico_id):
        EliminaEsitoDialog(self, medico_id, 'medico')

    def aggiorna_esito(self):
        self.populate_miei_esiti()
        self.populate_esiti()

    def modifica_esito(self, medico_id):
        ModificaEsitoDialog(self, medico_id, 'medico')

    def clear_frame(self):
        for widget in self.winfo_children():
            widget.destroy()

    def logout(self):
        # Mostra una finestra di dialogo di conferma
        risposta = messagebox.askyesno("Conferma Logout", "Sei sicuro di voler uscire?", parent = self)
        
        if risposta:  # Se l'utente conferma
            # Chiudi la connessione al database se è aperta
            if self.conn is not None and self.conn.is_connected():
                self.conn.close()
            
            # Chiudi la finestra corrente
            self.destroy()

class AggiungiAssenzaDialog(ctk.CTkToplevel):
    def __init__(self, parent, medico_id=None):
        super().__init__(parent)
        self.title("Aggiungi Assenza")
        self.geometry("500x350")

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        self.medico_id = medico_id
        self.medico_var = ctk.StringVar(self)

        # Connessione al database
        self.conn = connessione_db()

        if not self.medico_id:
            ctk.CTkLabel(self, text="Seleziona Medico").pack(pady=5)
            self.lista_medici = self.get_medici()
            self.medico_menu = ctk.CTkOptionMenu(self, variable=self.medico_var, values=list(self.lista_medici.keys()))
            self.medico_menu.pack(pady=5)

        ctk.CTkLabel(self, text="Data Inizio (YYYY-MM-DD)").pack(pady=5)
        self.entry_data_inizio = ctk.CTkEntry(self)
        self.entry_data_inizio.pack(pady=5)

        ctk.CTkLabel(self, text="Data Fine (YYYY-MM-DD)").pack(pady=5)
        self.entry_data_fine = ctk.CTkEntry(self)
        self.entry_data_fine.pack(pady=5)

        ctk.CTkButton(self, text="Aggiungi", command=self.aggiungi_assenza).pack(pady=20)

        # Attende che la finestra di dialogo venga chiusa prima di continuare
        self.wait_window(self)

    def aggiungi_assenza(self):
        if not self.medico_id:
            medico_selezionato = self.medico_var.get()
            if not medico_selezionato:
                messagebox.showerror("Errore", "Seleziona un medico", parent = self)
                return
            self.medico_id = self.lista_medici[medico_selezionato]

        data_inizio = self.entry_data_inizio.get()
        data_fine = self.entry_data_fine.get()

        if not data_inizio or not data_fine:
            messagebox.showerror("Errore", "Tutti i campi sono obbligatori.", parent = self)
            return

        try:
            data_inizio_dt = datetime.strptime(data_inizio, "%Y-%m-%d")
            data_fine_dt = datetime.strptime(data_fine, "%Y-%m-%d")
        except ValueError:
            messagebox.showerror("Errore", "Formato data non valido. Usa YYYY-MM-DD.", parent = self)
            return

        oggi_dt = datetime.today().date()
        if data_inizio_dt.date() < oggi_dt:
            messagebox.showerror("Errore", "L'assenza non può essere antecedente a oggi", parent = self)
            return

        if data_fine_dt < data_inizio_dt:
            messagebox.showerror("Errore", "La data di fine non può essere prima della data di inizio.", parent = self)
            return

        c = self.conn.cursor()

        # Controllare la sovrapposizione delle assenze
        c.execute("""
            SELECT * FROM assenze 
            WHERE medico = %s AND (
                (%s <= fine AND %s >= inizio) OR
                (%s <= fine AND %s >= inizio)
            )
        """, (self.medico_id, data_inizio, data_inizio, data_fine, data_fine))

        overlapping_assenze = c.fetchall()

        if overlapping_assenze:
            messagebox.showerror("Errore", "C'è già un'assenza in questo periodo.", parent = self)
            return

        # Inserimento della nuova assenza
        c.execute("INSERT INTO assenze (medico, inizio, fine) VALUES (%s, %s, %s)",
                  (self.medico_id, data_inizio, data_fine))
        self.conn.commit()

        messagebox.showinfo("Successo", "Assenza aggiunta con successo.", parent = self)
        self.destroy()

    def get_medici(self):
        c = self.conn.cursor()
        c.execute("SELECT ID, Nome, Cognome FROM medici")
        medici_rows = c.fetchall()
        return {f"{nome} {cognome}": id for id, nome, cognome in medici_rows}

    def __del__(self):
        """Chiude la connessione al database quando l'istanza viene distrutta."""
        if self.conn.is_connected():
            self.conn.close()

class EliminaAssenzaDialog(ctk.CTkToplevel):
    def __init__(self, parent, medico_id=None):
        super().__init__(parent)
        self.title("Elimina Assenza")
        self.geometry("500x350")

        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        self.medico_id = medico_id

        # Connessione al database
        self.conn = connessione_db()

        self.medico_var = ctk.StringVar(self)
        self.assenze = []

        if not self.medico_id:
            ctk.CTkLabel(self, text="Seleziona Medico").pack(pady=5)
            self.medico_dict = self.get_medici()
            self.medico_menu = ctk.CTkOptionMenu(self, variable=self.medico_var, values=list(self.medico_dict.keys()), command=self.update_medico_dict)
            self.medico_menu.pack(pady=5)

        self.listbox = tk.Listbox(self, height=10)
        self.listbox.pack(fill=ctk.BOTH, expand=True, padx=10, pady=10)

        ctk.CTkButton(self, text="Elimina", command=self.elimina_assenza).pack(pady=10)

       
        if self.medico_id:
            self.load_assenze()

    def get_medici(self):
        c = self.conn.cursor()
        c.execute("SELECT ID, Nome, Cognome FROM medici")
        medici = c.fetchall()

        medico_dict = {f"{nome} {cognome}": id for id, nome, cognome in medici}
        return medico_dict

    def update_medico_dict(self, selected_medico):
        """Aggiorna l'ID del medico selezionato e carica le assenze per il nuovo medico."""
        # Aggiorna il medico_id in base alla selezione corrente
        self.medico_id = self.medico_dict.get(selected_medico)

        # Chiama `load_assenze` per aggiornare la listbox con le assenze del nuovo medico
        self.load_assenze()

    def load_assenze(self):
        """Carica e visualizza le assenze del medico selezionato nella listbox."""
        # Ripulisce la listbox all'inizio della funzione
        self.listbox.delete(0, ctk.END)
        self.assenze = []

        # Recupera le assenze del medico selezionato
        c = self.conn.cursor()
        c.execute("SELECT id, inizio, fine FROM assenze WHERE medico = %s", (self.medico_id,))
        self.assenze = c.fetchall()

        # Inserisce un messaggio nella listbox se non ci sono assenze, altrimenti le mostra
        if not self.assenze:
            self.listbox.insert(ctk.END, "Nessuna assenza registrata")
        else:
            for assenza in self.assenze:
                data_inizio = assenza[1].strftime("%Y-%m-%d")
                data_fine = assenza[2].strftime("%Y-%m-%d")
                self.listbox.insert(ctk.END, f"{data_inizio} - {data_fine}")


    def elimina_assenza(self):

        selected = self.listbox.curselection()
        if not selected:
            messagebox.showerror("Errore", "Seleziona un'assenza da eliminare.", parent = self)
            return

        assenza_id = self.assenze[selected[0]][0]

        c = self.conn.cursor()
        c.execute("DELETE FROM assenze WHERE id = %s", (assenza_id,))
        self.conn.commit()

        messagebox.showinfo("Successo", "Assenza eliminata con successo.", parent = self)
        self.destroy()

    def __del__(self):
        """Chiude la connessione al database quando l'istanza viene distrutta."""
        if self.conn.is_connected():
            self.conn.close()

class ModificaAssenza(ctk.CTkToplevel):
    def __init__(self, parent, medico_id):
        super().__init__(parent)
        self.title("Gestione Assenze")
        self.geometry("300x400+700+250")
        self.medico_id = medico_id
        self.conn = connessione_db()

        self.transient(parent)

        self.create_widgets()
        self.populate_assenze_list()

    def create_widgets(self):
        # Mostra le assenze in un ListBox
        self.assenze_listbox = Listbox(self)
        self.assenze_listbox.pack(fill="both", expand=True)

        # Aggiungi un bottone per confermare la selezione e modificare l'assenza
        ctk.CTkButton(self, text="Modifica", command=self.modifica_assenza_selezionata).pack()

    def populate_assenze_list(self):
        c = self.conn.cursor()
        c.execute("SELECT id, inizio, fine FROM assenze WHERE medico = %s", (self.medico_id,))
        self.assenze = c.fetchall()
        c.close()

        for idx, (id, inizio, fine) in enumerate(self.assenze):
            self.assenze_listbox.insert(idx, f"ID {id}: Dal {inizio} al {fine}")

    def modifica_assenza_selezionata(self):
        selected_index = self.assenze_listbox.curselection()
        if not selected_index:
            messagebox.showerror("Errore", "Seleziona un'assenza", parent=self)
            return

        # Ottieni l'ID dell'assenza selezionata
        assenza_id, inizio, fine = self.assenze[selected_index[0]]
        self.apri_finestra_modifica(assenza_id, inizio, fine)

    def apri_finestra_modifica(self, assenza_id, inizio, fine):
        edit_window = ctk.CTkToplevel(self)
        edit_window.title("Modifica Assenza")
        edit_window.geometry("400x300+700+250")

        edit_window.attributes('-topmost', True)

        lbl_inizio = ctk.CTkLabel(edit_window, text="Inizio (YYYY-MM-DD)")
        lbl_inizio.pack()
        entry_inizio = ctk.CTkEntry(edit_window)
        entry_inizio.insert(0, inizio)
        entry_inizio.pack()

        lbl_fine = ctk.CTkLabel(edit_window, text="Fine (YYYY-MM-DD)")
        lbl_fine.pack()
        entry_fine = ctk.CTkEntry(edit_window)
        entry_fine.insert(0, fine)
        entry_fine.pack()

        ctk.CTkButton(edit_window, text="Salva Modifiche",
                      command=lambda: self.salva_modifiche_assenza(assenza_id, entry_inizio.get(), entry_fine.get(), edit_window)).pack()

    def salva_modifiche_assenza(self, assenza_id, nuovo_inizio, nuova_fine, edit_window):
        try:
            data_inizio_dt = datetime.strptime(nuovo_inizio, "%Y-%m-%d")
            data_fine_dt = datetime.strptime(nuova_fine, "%Y-%m-%d")

            oggi_dt = datetime.today().date()
            if data_inizio_dt.date() < oggi_dt:
                messagebox.showerror("Errore", "L'assenza non può essere antecedente a oggi", parent=edit_window)
                return

            if data_fine_dt < data_inizio_dt:
                messagebox.showerror("Errore", "La data di fine non può essere prima della data di inizio.", parent=edit_window)
                return

            c = self.conn.cursor()

            # Controllare la sovrapposizione delle assenze
            c.execute("""
                SELECT * FROM assenze 
                WHERE medico = %s AND (
                    (%s <= fine AND %s >= inizio) OR
                    (%s <= fine AND %s >= inizio)
                )
            """, (self.medico_id, nuovo_inizio, nuovo_inizio, nuova_fine, nuova_fine))

            overlapping_assenze = c.fetchall()
            if overlapping_assenze:
                messagebox.showerror("Errore", "C'è già un'assenza in questo periodo.", parent=edit_window)
                return

            # Aggiornare i dati dell'assenza
            c.execute("UPDATE assenze SET inizio = %s, fine = %s WHERE id = %s", (nuovo_inizio, nuova_fine, assenza_id))
            self.conn.commit()

            messagebox.showinfo("Successo", "Assenza modificata correttamente.", parent=edit_window)
            edit_window.destroy()

        except ValueError as e:
            messagebox.showerror("Errore", f"Errore formato data: {str(e)}", parent=edit_window)

class InserisciEsitoDialog(ctk.CTkToplevel):
    def __init__(self, parent, medico_id=None):
        super().__init__(parent)
        self.medico_id = medico_id
        self.title("Inserisci Esito")
        self.geometry("900x530+300+100")

        self.conn = connessione_db()

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        if not self.medico_id:
            ctk.CTkLabel(self, text="Seleziona Medico").grid(row=0, column=0, pady=5, padx=5, sticky='w')
            self.medico_var = ctk.StringVar(self)
            self.medico_dict = self.get_medici()
            self.medico_menu = ctk.CTkOptionMenu(self, values=list(self.medico_dict.keys()), variable=self.medico_var)
            self.medico_menu.grid(row=0, column=1, pady=5, padx=5, sticky='ew')
        ctk.CTkLabel(self, text="Seleziona Paziente").grid(row=0, column=2, pady=5, padx=5, sticky='w')
        self.paziente_var = ctk.StringVar(self)
        self.lista_pazienti = self.get_pazienti()


        # Creazione del CTkOptionMenu con colori personalizzati
        self.paziente_menu = ctk.CTkOptionMenu(
            self, 
            values=list(self.lista_pazienti.keys()), 
            variable=self.paziente_var,
            text_color = "black",
            fg_color="lightgray",  # Colore di sfondo del menu
            button_color="orange",  # Colore del pulsante
            button_hover_color="orange",  # Colore del pulsante al passaggio del mouse
            dropdown_fg_color="white",  # Colore di sfondo delle opzioni
            dropdown_hover_color="lightgray"  # Colore delle opzioni al passaggio del mouse
        )
        self.paziente_menu.grid(row=0, column=3, pady=5, padx=5, sticky='ew')


        # Scelta dell'ambulatorio
        ctk.CTkLabel(self, text="Ambulatorio").grid(row=1, column=0, pady=5, padx=5, sticky='w')
        self.ambulatorio_var = ctk.StringVar(self)
        self.ambulatorio_menu = ctk.CTkOptionMenu(
            self, 
            values=["Pediatrico", "Per Adulti"], 
            variable=self.ambulatorio_var,
            fg_color="lightgray",             # Colore di sfondo del menu
            text_color = "black",
            button_color="orange",         # Colore del pulsante
            button_hover_color="orange", # Colore del pulsante al passaggio del mouse
            dropdown_fg_color="white",   # Colore di sfondo delle opzioni
            dropdown_hover_color="lightgray" # Colore delle opzioni al passaggio del mouse
        )
        self.ambulatorio_menu.grid(row=1, column=1, pady=5, padx=5, sticky='ew')

        ctk.CTkLabel(self, text="Urgenza").grid(row=1, column=2, pady=5, padx=5, sticky='w')
        self.urgenza_var = ctk.BooleanVar()
        self.urgenza_checkbox = ctk.CTkCheckBox(self, text="Sì", variable=self.urgenza_var, command=self.handle_urgency)
        self.urgenza_checkbox.grid(row=1, column=3, pady=5, padx=5, sticky='w')

        ctk.CTkLabel(self, text="Tipo di Visita").grid(row=2, column=0, pady=5, padx=5, sticky='w')
        self.tipo_visita_var = ctk.StringVar(self)
        self.tipo_visita_menu = ctk.CTkOptionMenu(self, values=["Medico curante", "Sostituto", "Urgenza"], variable=self.tipo_visita_var)
        self.tipo_visita_menu.grid(row=2, column=1, pady=5, padx=5, sticky='ew')

        # Aggiunta di un campo per le note
        ctk.CTkLabel(self, text="Referto visita").grid(row=3, column=0, pady=5, padx=5, columnspan=4)
        self.note_var = tk.StringVar(self)
        note_entry = ctk.CTkTextbox(self, height=300, width=890)
        note_entry.grid(row=4, column=0, pady=5, padx=5, columnspan=4, sticky='ew')

        # Puoi legare il contenuto del Textbox alla variabile note_var
        note_entry.bind("<KeyRelease>", lambda event: self.note_var.set(note_entry.get("1.0", "end-1c")))

        ctk.CTkButton(self, text="Invia", command=lambda: self.inserisci_esito(self.urgenza_checkbox)).grid(row=5, column=0, pady=10, padx=5, columnspan=4)

    def handle_urgency(self):
        if self.urgenza_var.get():
            self.tipo_visita_var.set("Urgenza")
            self.tipo_visita_menu.configure(state='disabled')
        else:
            self.tipo_visita_menu.configure(state='normal')

    def get_medici(self):
        
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT ID, Nome, Cognome FROM medici")
        medici = c.fetchall()
        self.conn.close()
        return {f"{nome} {cognome}": id for id, nome, cognome in medici}
        
    def get_pazienti(self):
        
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT ID, Nome, Cognome FROM pazienti ",)
        pazienti = c.fetchall()
        self.conn.close()
        return {f"{nome} {cognome}": id for id, nome, cognome in pazienti}

    def inserisci_esito(self, urgenza):

        if not self.medico_id:
            medico_selezionato = self.medico_var.get()
            if not medico_selezionato:
                messagebox.showerror("Errore", "Seleziona un infermiere", parent = self)
                return

            medico_id = self.medico_dict[medico_selezionato]
        else:
            medico_id = self.medico_id

        if not medico_id:
            messagebox.showerror("Errore", "Seleziona un medico", parent = self)
            return
        
        paziente_selezionato = self.paziente_var.get()
        if not paziente_selezionato:
            messagebox.showerror("Errore", "Seleziona un paziente", parent = self)
            return
        
        # Controlla che l'ambulatorio sia stato selezionato
        ambulatorio = self.ambulatorio_var.get()
        if not ambulatorio:
            messagebox.showerror("Errore", "Seleziona un ambulatorio.", parent = self)
            return
        
        # Controlla che la visita sia stato selezionato
        tipo_visita = self.tipo_visita_var.get()
        if not tipo_visita:
            messagebox.showerror("Errore", "Seleziona il tipo di visita.", parent = self)
            return
        
        # Controlla che ci sia scritto qualcosa nel referto
        esito_messaggio = self.note_var.get()
        if not esito_messaggio:
            messagebox.showerror("Errore", "Compila il referto.", parent = self)
            return
        
        urgenza = self.urgenza_var.get() if not urgenza else "Urgenza"
        paziente_id = self.lista_pazienti[paziente_selezionato]

        # Ottieni la data e l'ora corrente
        data_corrente = datetime.now().strftime("%Y-%m-%d")

        # Inserisci l'esito nel database
        self.inserisci_esito_nel_database(medico_id, paziente_id, esito_messaggio, ambulatorio, data_corrente, urgenza, tipo_visita)

        messagebox.showinfo("Successo", "Esito inserito con successo", parent = self)
        self.destroy()

    def inserisci_esito_nel_database(self, medico_id, paziente_id, esito, ambulatorio, data, urgenza, tipo_visita):
        self.conn = connessione_db()
        c = self.conn.cursor()
        c.execute("""
            INSERT INTO esiti (medico_id, paziente_id, esito, ambulatorio, data, urgenza, tipo_visita)
            VALUES (%s, %s, %s, %s, %s, %s, %s)
        """, (medico_id, paziente_id, esito, ambulatorio, data, urgenza, tipo_visita))
        self.conn.commit()
        c.close()

class EliminaEsitoDialog(ctk.CTkToplevel):
    def __init__(self, parent, user_id, user_type):
        super().__init__(parent)
        self.title(f"Elimina Esito ({'Medico' if user_type == 'medico' else 'Infermiere' if user_type == 'infermiere' else 'Segreteria'})")
        self.geometry("900x450")
        self.transient(parent)
        self.user_id = user_id
        self.user_type = user_type

        self.conn = connessione_db()
        self.create_widgets()

    def create_widgets(self):
        ctk.CTkLabel(self, text="Seleziona un esito dalla lista:").pack(pady=5)

        self.listbox_esiti = tk.Listbox(self, width=100)
        self.listbox_esiti.pack(pady=5, fill="both", expand=True)
        self.listbox_esiti.bind('<<ListboxSelect>>', self.seleziona_esito)

        ctk.CTkButton(self, text="Elimina Esito", command=self.elimina_esito_selezionato).pack(pady=20)

        self.carica_esiti()

    def carica_esiti(self):
        c = self.conn.cursor()
        if self.user_type == 'segreteria':
            query = "SELECT id, data, esito FROM esiti"
            c.execute(query)
        else:
            query = "SELECT id, data, esito FROM esiti WHERE {}_id = %s".format(self.user_type)
            c.execute(query, (self.user_id,))
        self.esiti = c.fetchall()
        c.close()

        for esito in self.esiti:
            esito_str = f"ID {esito[0]} - Data: {esito[1]} - Esito: {esito[2]}"
            self.listbox_esiti.insert(tk.END, esito_str)

    def seleziona_esito(self, event):
        index = self.listbox_esiti.curselection()
        if index:
            self.id_selezionato = self.esiti[index[0]][0]

    def elimina_esito_selezionato(self):
        if not hasattr(self, 'id_selezionato'):
            messagebox.showerror("Errore", "Seleziona un esito dalla lista.", parent=self)
            return

        response = messagebox.askyesno("Conferma", "Sei sicuro di voler eliminare questo esito?", parent=self)
        if response:
            c = self.conn.cursor()
            c.execute("DELETE FROM esiti WHERE id = %s", (self.id_selezionato,))
            self.conn.commit()
            c.close()
            messagebox.showinfo("Successo", "Esito eliminato con successo.", parent=self)
            self.destroy()  # Chiude la finestra dopo l'eliminazione

class ModificaEsitoDialog(ctk.CTkToplevel):
    def __init__(self, parent, user_id, user_type):
        super().__init__(parent)
        self.title("Modifica Esito")
        self.geometry("1100x500")
        self.transient(parent)
        self.user_id = user_id
        self.user_type = user_type
        self.conn = connessione_db()

        self.selected_esito = None  # Inizializza la variabile per tenere traccia dell'esito selezionato
        self.create_widgets()
        self.populate_esiti_list()

    def create_widgets(self):
        ctk.CTkLabel(self, text="Seleziona un esito dalla lista:").pack(pady=5)

        self.listbox_esiti = tk.Listbox(self, width=100)
        self.listbox_esiti.pack(pady=5, fill="both", expand=True)
        self.listbox_esiti.bind('<<ListboxSelect>>', self.seleziona_esito)

        ctk.CTkButton(self, text="Modifica", command=self.apri_finestra_modifica).pack(pady=20)

    def populate_esiti_list(self):
        c = self.conn.cursor()
        if self.user_type == 'segreteria':
            query = "SELECT id, data, esito, ambulatorio, urgenza, tipo_visita, paziente_id FROM esiti ORDER BY data DESC"
        else:
            query = "SELECT id, data, esito, ambulatorio, urgenza, tipo_visita, paziente_id FROM esiti WHERE {}_id = %s ORDER BY data DESC".format(self.user_type)
            c.execute(query, (self.user_id,))
        self.esiti = c.fetchall()
        c.close()

        for esito in self.esiti:
            esito_str = f"ID {esito[0]} - Data: {esito[1]} - Esito: {esito[2]} - Ambulatorio: {esito[3]} - Urgenza: {esito[4]} - Tipo: {esito[5]}"
            self.listbox_esiti.insert(tk.END, esito_str)

    def seleziona_esito(self, event):
        index = self.listbox_esiti.curselection()
        if index:
            self.selected_esito = self.esiti[index[0]]
        else:
            self.selected_esito = None  # Resetta la selezione se non valida

    def apri_finestra_modifica(self):
        if self.selected_esito:
            EditEsitoDialog(self, self.selected_esito)
        else:
            messagebox.showerror("Errore", "Nessun esito selezionato.", parent=self)

class EditEsitoDialog(ctk.CTkToplevel):
    def __init__(self, parent, esito):
        super().__init__(parent)
        self.title("Modifica Dettagli Esito")
        self.geometry("1100x600")  # Dimensione adeguata per campi aggiuntivi
        self.transient(parent)
        self.esito = esito

        self.conn = connessione_db()
        self.create_form()
        self.create_buttons()

    def create_form(self):
        # Entry per la data
        ctk.CTkLabel(self, text="Data (YYYY-MM-DD):").pack()
        self.entry_data = ctk.CTkEntry(self)
        self.entry_data.insert(0, self.esito[1])
        self.entry_data.pack()

        # Entry per ambulatorio
        ctk.CTkLabel(self, text="Ambulatorio:").pack()
        self.entry_ambulatorio = ctk.CTkEntry(self)
        self.entry_ambulatorio.insert(0, self.esito[3])
        self.entry_ambulatorio.pack()

        # Entry per urgenza
        ctk.CTkLabel(self, text="Urgenza (0=No, 1=Si):").pack()
        self.entry_urgenza = ctk.CTkEntry(self)
        self.entry_urgenza.insert(0, self.esito[4])
        self.entry_urgenza.pack()

        # Entry per tipo di visita
        ctk.CTkLabel(self, text="Tipo Visita:").pack()
        self.entry_tipo_visita = ctk.CTkEntry(self)
        self.entry_tipo_visita.insert(0, self.esito[5])
        self.entry_tipo_visita.pack()

        # Entry per esito
        ctk.CTkLabel(self, text="Esito:").pack()
        self.entry_esito = ctk.CTkEntry(self, width=900)  # Larghezza modificata
        self.entry_esito.insert(0, self.esito[2])
        self.entry_esito.pack(fill='x', padx=20)

        # Combo box per selezione paziente
        ctk.CTkLabel(self, text="Paziente:").pack()
        self.combo_paziente_id = ttk.Combobox(self, width=90)
        self.load_pazienti()  # Carica l'elenco dei pazienti
        self.combo_paziente_id.set(f"{self.esito[6]} - {self.get_paziente_nome(self.esito[6])}")
        self.combo_paziente_id.pack()

    def load_pazienti(self):
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM pazienti")
        pazienti = c.fetchall()
        self.pazienti_dict = {f"{id} - {nome} {cognome}": id for id, nome, cognome in pazienti}
        self.combo_paziente_id['values'] = list(self.pazienti_dict.keys())

    def get_paziente_nome(self, paziente_id):
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT nome, cognome FROM pazienti WHERE id = %s", (paziente_id,))
        nome, cognome = c.fetchone()
        return f"{nome} {cognome}"

    def create_buttons(self):
        save_button = ctk.CTkButton(self, text="Salva Modifiche", command=self.salva_modifiche)
        save_button.pack(pady=10)

    def salva_modifiche(self):
        try:
            # Convalida il formato della data
            data = self.entry_data.get()
            datetime.strptime(data, "%Y-%m-%d")  # Se la data non è valida, solleva ValueError

            ambulatorio = self.entry_ambulatorio.get()
            urgenza = self.entry_urgenza.get()
            tipo_visita = self.entry_tipo_visita.get()
            esito = self.entry_esito.get()
            paziente_id = self.pazienti_dict[self.combo_paziente_id.get()]

            # Esegue l'aggiornamento del database
            c = self.conn.cursor()
            c.execute("""
                UPDATE esiti SET 
                data = %s, ambulatorio = %s, urgenza = %s, tipo_visita = %s, esito = %s, paziente_id = %s
                WHERE id = %s
            """, (data, ambulatorio, urgenza, tipo_visita, esito, paziente_id, self.esito[0]))
            self.conn.commit()
            messagebox.showinfo("Successo", "Esito aggiornato con successo!", parent=self)
            self.destroy()

        except ValueError as e:
            # Gestisce l'errore se il formato della data non è corretto
            messagebox.showerror("Errore", f"Formato data non valido. Usa YYYY-MM-DD.\nDettagli errore: {e}", parent=self)
## fine medico

# infermiere
class InfermiereInterface(ctk.CTkToplevel):
    def __init__(self, parent, infermiere,email, password):
        super().__init__(parent)
        self.title("Interfaccia Infermiere")
        self.geometry("1200x600+70+50")
        self.infermiere = infermiere
        self.email = email
        self.infermiere_id = infermiere.get('id')
        self.nome = infermiere.get('nome')
        self.cognome = infermiere.get('cognome')
        self.password = password

        self.create_widgets()

    def create_widgets(self):

        self.clear_frame()

        self.conn = connessione_db()
        c = self.conn.cursor()
        c.execute("SELECT nome, cognome FROM infermieri WHERE email = %s AND password = %s", (self.email, self.password))
        infermiere_data = c.fetchone()
        self.conn.close()

        nome, cognome = infermiere_data

        # Frame superiore per nome e cognome 
        frame_top = ctk.CTkFrame(self, fg_color="#59B7D8", height=50)
        frame_top.pack(fill="x")

        # Nome e cognome 
        nome_label = ctk.CTkLabel(frame_top, text=f"Inf. {nome} {cognome}", font=("Long Island", 20))
        nome_label.grid(row=0, column=0, padx=(10, 5), pady=10, sticky="w")

        # Creazione del bottone per il logout
        logout_button = ctk.CTkButton(frame_top, text="Logout", command=self.logout)
        logout_button.grid(row=0, column=1, padx=10, pady=10, sticky="e")

        # Allinea tutto in una griglia con lo spazio rimanente
        frame_top.grid_columnconfigure(0, weight=1)

        # Frame inferiore per i bottoni
        frame_bottom = ctk.CTkFrame(self, height = 50, fg_color="#59B7D8")
        frame_bottom.pack(side="bottom", fill="x")

        # Frame sinistro per il notebook principale
        frame_left = ctk.CTkFrame(self, width=450, fg_color="white")
        frame_left.pack(side="left", fill="both", expand=False)
        frame_left.pack_propagate(False)

        # Creazione del Tabview principale
        notebook_main = ctk.CTkTabview(frame_left)
        notebook_main.pack(fill="both", expand=True, padx=10, pady=10)

        # Aggiunta della Tab "Pazienti"
        notebook_main.add("Pazienti")
        tab_pazienti = notebook_main.tab("Pazienti")

        # Tab "Pazienti"
        frame_bottoni_pazienti = ctk.CTkFrame(tab_pazienti)
        frame_bottoni_pazienti.pack(side="bottom", pady=10)

        # Pulsante di modifica del paziente all'interno del frame
        ctk.CTkButton(frame_bottoni_pazienti, text="Modifica Paziente", command=lambda: self.apri_modifica_dialog("pazienti")).pack(side = "left", padx=10)
        ctk.CTkButton(frame_bottoni_pazienti, text="Aggiorna", command=self.populate_tutti_pazienti).pack(side = "left", padx=10)

        # Crea e configura la Listbox
        self.listbox_pazienti = tk.Listbox(tab_pazienti, height= 50)
        self.listbox_pazienti.pack(side='top', fill='x')
        self.populate_tutti_pazienti()

        # Aggiunta della Tab "Visite, Prelievi e Medicazioni"
        notebook_main.add("Visite, Prelievi e Medicazioni")
        tab_visite = notebook_main.tab("Visite, Prelievi e Medicazioni")
        sub_notebook_visite = ctk.CTkTabview(tab_visite)
        sub_notebook_visite.pack(fill="both", expand=True)

        sub_tab_miei_esiti = sub_notebook_visite.add("Miei esiti")
        self.textbox_miei_esiti = ctk.CTkTextbox(sub_tab_miei_esiti, width=200, wrap="word")
        self.textbox_miei_esiti.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_miei_esiti.configure(state="disabled") # Rende il textbox non modificabile dall'utente
        self.populate_miei_esiti() 

        sub_tab_tutti_esiti = sub_notebook_visite.add("Tutti gli esiti")
        self.textbox_tutti_esiti = ctk.CTkTextbox(sub_tab_tutti_esiti, width=200, wrap="word")
        self.textbox_tutti_esiti.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutti_esiti.configure(state="disabled") # Rende il textbox non modificabile dall'utente
        self.populate_esiti() 

        # Frame centrale per i dettagli del paziente
        self.frame_center = ctk.CTkFrame(self, width = 350, fg_color="white")
        self.frame_center.pack(side="left", fill="both", expand=True)
        self.frame_center.pack_propagate(False)

        # Frame destro per "Prenotazioni"
        frame_right = ctk.CTkFrame(self, width=450, fg_color="white")
        frame_right.pack(side="right", fill="both")
        frame_right.pack_propagate(False)
        ctk.CTkLabel(frame_right, text="Prenotazioni", font=("Long Island", 16)).pack(pady=10)

        # Creazione del Tabview per "Prenotazioni"
        notebook_comunicazioni = ctk.CTkTabview(frame_right)
        notebook_comunicazioni.pack(fill="both", expand=True)

        # Tab per "Storico"
        tab_mie_comunicazioni = notebook_comunicazioni.add("Storico")
        self.textbox_mie_comunicazioni = ctk.CTkTextbox(tab_mie_comunicazioni, width=200, height=350, wrap="word")
        self.textbox_mie_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_mie_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        # Tab per "Prossime prenotazioni"
        tab_tutte_comunicazioni = notebook_comunicazioni.add("Prossime prenotazioni")
        self.textbox_tutte_comunicazioni = ctk.CTkTextbox(tab_tutte_comunicazioni, width=200, height=350, wrap="word")
        self.textbox_tutte_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutte_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        # Frame per il bottone Aggiorna nel frame Prenotazioni
        frame_bottone_prenotazioni = ctk.CTkFrame(frame_right, width=900, height=100, fg_color="white")
        frame_bottone_prenotazioni.pack(side="bottom")
        frame_bottone_prenotazioni.pack_propagate(False)
        ctk.CTkButton(frame_bottone_prenotazioni, text="Aggiorna", command=self.aggiorna_comunicazioni_infermiere).pack() 

        ctk.CTkButton(frame_bottom, text="Inserisci Esito", command=self.apri_finestra_inserisci_esito).pack(side="left", padx=10, pady=10)
        ctk.CTkButton(frame_bottom, text="Calendario", command=lambda: self.visualizza_calendario(self.infermiere_id)).pack(side="right", padx=10, pady=10)

        # Creazione di un frame per i bottoni nella tab 'visite'
        frame_bottoni_visite = ctk.CTkFrame(tab_visite, height = 200)
        frame_bottoni_visite.pack(side="bottom", pady=10)
        frame_bottoni_visite.pack_propagate(False)

        # Inserimento di bottoni in una griglia 2x2
        ctk.CTkButton(frame_bottoni_visite, text="Elimina Esito", command=lambda: self.elimina_esito(self.infermiere_id)).grid(row=0, column=0, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_visite, text="Aggiorna", command=self.aggiorna_esito).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        ctk.CTkButton(frame_bottoni_visite, text="Modifica Esito", command=lambda:self.modifica_esito(self.infermiere_id)).grid(row=1, column=0, padx=10, pady=10, sticky="ew")

        # Configurazione della griglia per distribuire uniformemente lo spazio
        frame_bottoni_visite.grid_columnconfigure(0, weight=1)
        frame_bottoni_visite.grid_columnconfigure(1, weight=1)

        # Per assicurarsi che i bottoni riempiano lo spazio disponibile, è possibile configurare il frame per espandere
        frame_bottoni_visite.pack(fill="x")

        # Aggiornare le comunicazioni all'avvio
        self.aggiorna_comunicazioni_infermiere() # da rivedere
    
    def apri_modifica_dialog(self, tipo):
        ModificaDatiDialog(self, tipo)

    def aggiorna_comunicazioni_infermiere(self):

        oggi = datetime.today().date()
        conn = connessione_db()
        c = conn.cursor()

        # Recuperare le prenotazioni del medico che sono antecedenti a oggi
        c.execute("""
            SELECT p.data, p.orario, p.sala, pa.Nome, pa.Cognome
            FROM prenotazioni p
            JOIN infermieri i ON p.infermiere_id = i.id
            JOIN pazienti pa ON p.paziente_id = pa.id
            WHERE p.infermiere_id = %s AND p.data < %s
            ORDER BY p.data DESC, p.orario
        """, (self.infermiere_id, oggi))
        mie_prenotazioni_passate = c.fetchall()

        # Recuperare le prenotazioni future del medico
        c.execute("""
            SELECT p.data, p.orario, p.sala, pa.Nome, pa.Cognome, i.Nome, i.Cognome
            FROM prenotazioni p
            JOIN infermieri i ON p.infermiere_id = i.id
            JOIN pazienti pa ON p.paziente_id = pa.id
            WHERE p.infermiere_id = %s AND p.data >= %s
            ORDER BY p.data, p.orario
        """, (self.infermiere_id, oggi))
        tutte_prenotazioni_future = c.fetchall()
        
        c.close()

        # Aggiornare il contenuto delle etichette con le prenotazioni
        mie_prenotazioni_passate_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')} alle {orario} in {sala} con {nome} {cognome}" for data, orario, sala, nome, cognome in mie_prenotazioni_passate]
        ) if mie_prenotazioni_passate else "Nessuna prenotazione registrata."

        tutte_prenotazioni_future_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')} alle {orario} in {sala} con paziente {nome_paziente} {cognome_paziente} e infermiere {nome_infermiere} {cognome_infermiere}" 
            for data, orario, sala, nome_paziente, cognome_paziente, nome_infermiere, cognome_infermiere in tutte_prenotazioni_future]
        ) if tutte_prenotazioni_future else "Nessuna prenotazione."

        # Aggiornare il contenuto delle Textbox
        self.textbox_mie_comunicazioni.configure(state="normal")
        self.textbox_mie_comunicazioni.delete("1.0", "end")
        self.textbox_mie_comunicazioni.insert("1.0", mie_prenotazioni_passate_str)
        self.textbox_mie_comunicazioni.configure(state="disabled")

        self.textbox_tutte_comunicazioni.configure(state="normal")
        self.textbox_tutte_comunicazioni.delete("1.0", "end")
        self.textbox_tutte_comunicazioni.insert("1.0", tutte_prenotazioni_future_str)
        self.textbox_tutte_comunicazioni.configure(state="disabled")

    def elimina_esito(self, infermiere_id):
        EliminaEsitoDialog(self, infermiere_id, 'infermiere')

    def aggiorna_esito(self):
        self.populate_miei_esiti()
        self.populate_esiti()

    def modifica_esito(self, infermiere_id):
        ModificaEsitoDialog(self, infermiere_id, 'infermiere')

    def apri_finestra_inserisci_esito(self):
        InserisciEsitoInfDialog(self, self.infermiere_id, f"{self.nome} {self.cognome}")

    def populate_miei_esiti(self):

        # Pulisci la Textbox prima di inserire i nuovi dati
        self.textbox_miei_esiti.configure(state="normal")
        self.textbox_miei_esiti.delete("1.0", "end")

        conn = connessione_db()
        c = conn.cursor()
        c.execute("""
                SELECT e.esito, e.data, 
                CONCAT(p.nome, ' ', p.cognome) AS nome_completo_paziente
                FROM esiti e
                LEFT JOIN pazienti p ON e.paziente_id = p.id
                WHERE e.infermiere_id = %s
                ORDER BY e.data DESC
        """, (self.infermiere_id,))
        esiti = c.fetchall()
        c.close()

        # Inserisci i nuovi dati
        if esiti:
            for esito, data, nome_completo_paziente in esiti:
                self.textbox_miei_esiti.insert("end", f"- {data}\nPaziente: {nome_completo_paziente}\nEsito : {esito}\n\n")
        else:
            self.textbox_miei_esiti.insert("end", "Nessun esito registrato.")

        # Imposta lo stato di 'disabled' per evitare modifiche
        self.textbox_miei_esiti.configure(state="disabled")

    def populate_esiti(self):
        
        # Pulisci la Textbox prima di inserire i nuovi dati
        self.textbox_tutti_esiti.configure(state="normal")
        self.textbox_tutti_esiti.delete("1.0", "end")

        conn = connessione_db()
        c = conn.cursor()
        c.execute("""
            SELECT e.esito, e.data, e.urgenza, e.tipo_visita,
                COALESCE(m.Nome, i.Nome, s.Nome) AS Nome,
                COALESCE(m.Cognome, i.Cognome, s.Cognome) AS Cognome
            FROM esiti e
            LEFT JOIN medici m ON e.medico_id = m.id
            LEFT JOIN infermieri i ON e.infermiere_id = i.id
            LEFT JOIN segreteria s ON e.segreteria_id = s.id
            ORDER BY e.data DESC
        """)
        # Si utilizza COALESCE per ottenere il primo valore non nullo tra i campi nome e cognome delle tre tabelle

        esiti = c.fetchall()
        c.close()

        # Inserisci i nuovi dati
        if esiti:
            for esito, data, urgenza, tipo_visita, nome, cognome in esiti:
                self.textbox_tutti_esiti.insert("end", f"- {data} \tUrgenza: {urgenza} \nTipo visita: {tipo_visita}\nCaricato da: {nome} {cognome}\nEsito: {esito}\n\n")
        else:
            self.textbox_tutti_esiti.insert("end", "Nessun esito registrato.")

        # Imposta lo stato di 'disabled' per evitare modifiche
        self.textbox_tutti_esiti.configure(state="disabled")

    def populate_tutti_pazienti(self):

        self.listbox_pazienti.delete(0, tk.END)

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM pazienti Order by cognome")
        pazienti = c.fetchall()
        conn.close()

        for paziente in pazienti:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_pazienti.insert(tk.END, f"{paziente[0]} {paziente[2]} {paziente[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_pazienti.bind('<<ListboxSelect>>', self.seleziona_utente)

    def seleziona_utente(self, event):
        index = self.listbox_pazienti.curselection()
        if index:
            self.id_selezionato = self.listbox_pazienti.get(index[0]).split(" - ")[0]
            self.show_patient_details(self.id_selezionato)

    def show_patient_details(self, patient_id):
        for widget in self.frame_center.winfo_children():
            widget.destroy()

        conn = connessione_db()
        # Configura il cursore per restituire dizionari
        c = conn.cursor()
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico ,r.minorenne_id
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id 
                LEFT JOIN responsabilita r ON p.id = r.maggiorenne_id
                WHERE p.id = %s
            """, (patient_id,))
        paziente = c.fetchone()
        conn.close()

        if paziente:
            ctk.CTkLabel(self.frame_center, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Password: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(self.frame_center, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=5)


            # Controlla se il paziente ha minorenni a carico
            if paziente[11] is not None:
                ctk.CTkButton(self.frame_center, text="Visualizza Responsabilità", command=lambda: self.visualizza_responsabilita(paziente[0])).pack(pady=10)
        else:
            ctk.CTkLabel(self.frame_center, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

    def show_history(self, id = None):
        self.clear_frame()

        # Utilizzo di CTkTextbox al posto di Listbox
        self.text_box = ctk.CTkTextbox(self, height=10, wrap="word", state="normal", yscrollcommand=True)
        self.text_box.pack(pady=20, fill="both", expand=True)

        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        conn = connessione_db()
        cursor = conn.cursor()

        if id: 
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                WHERE p.id = %s
                ORDER BY e.data DESC
            """, (id,))
        else:
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                ORDER BY e.data DESC
            """)
        rows = cursor.fetchall()
        conn.close()

        self.text_box.delete('1.0', tk.END)  # Pulisce il textbox prima di inserire nuovi dati
        if not rows:  # Controlla se la lista è vuota
            self.text_box.insert(tk.END, "Nessun esito registrato")  # Mostra messaggio se non ci sono dati
        else:
            for row in rows:
                urgenza_status = "No" if row[2] == 0 else "Yes"
                healthcare_provider = f"Medico: {row[5]} {row[6]}" if row[5] and row[6] else f"Infermiere: {row[7]} {row[8]}"
                formatted_row = (f"- {row[0]} \t Ambulatorio: {row[1]} \tUrgenza: {urgenza_status} \t"
                                f"Regime della visita: {row[3]} \n"
                                f"{healthcare_provider} \n"
                                f"Paziente: {row[9]} {row[10]}\nEsito: {row[4]} \n\n")
                self.text_box.insert(tk.END, formatted_row) 

    def visualizza_responsabilita(self, id):
        self.clear_frame()  # Pulisce i widget esistenti

        # Bottone per tornare alla vista principale
        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        frame_responsabilita = ctk.CTkFrame(self, width=350)
        frame_responsabilita.pack(side="left", fill="both", expand=True)
        frame_responsabilita.pack_propagate(False)

        # Connessione al database MySQL
        conn = connessione_db()
        c = conn.cursor()  # Usare il cursore che ritorna dizionari

        # Query per recuperare i dettagli del paziente maggiorenne e il conteggio dei minorenni a carico
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico,
                    (SELECT COUNT(*) FROM responsabilita WHERE maggiorenne_id = p.id) AS count_minorenni
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id
                WHERE p.id = %s
            """, (id,))
        paziente = c.fetchone()

        # Query per ottenere i dettagli di ogni minorenne a carico
        c.execute("""
                SELECT p.*
                FROM pazienti p
                JOIN responsabilita r ON p.id = r.minorenne_id
                WHERE r.maggiorenne_id = %s
            """, (id,))
        minorenni = c.fetchall()

        # Mostra i dettagli del maggiorenne
        if paziente:
            ctk.CTkLabel(frame_responsabilita, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(frame_responsabilita, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=5)

            # Mostra i dettagli di ogni minorenne
            for minorenne in minorenni:
                min_frame = ctk.CTkFrame(self, width=350)
                min_frame.pack(side = "left",  fill="both", expand=True)
                min_frame.pack_propagate(False)
                ctk.CTkLabel(min_frame, text=f"Minorenne - Nome: {minorenne[1]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Cognome: {minorenne[2]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"CF: {minorenne[3]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Data di nascita: {minorenne[5]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Luogo di nascita: {minorenne[6]}").pack(pady=5)
                ctk.CTkButton(min_frame, text="Storico", command=lambda min_id=minorenne[0]: self.show_history(min_id)).pack(pady=5)

        else:
            ctk.CTkLabel(frame_responsabilita, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

        conn.close()

    def visualizza_calendario(self, infermiere_id):
        VisualizzaCalendario(self, infermiere_id)

    def clear_frame(self):
        for widget in self.winfo_children():
            widget.destroy()

    def logout(self):
        # Mostra una finestra di dialogo di conferma
        risposta = messagebox.askyesno("Conferma Logout", "Sei sicuro di voler uscire?", parent = self)
        
        if risposta:  # Se l'utente conferma
            # Chiudi la connessione al database se è aperta
            if self.conn is not None and self.conn.is_connected():
                self.conn.close()
            
            # Chiudi la finestra corrente
            self.destroy()

class VisualizzaCalendario(ctk.CTkToplevel):
    def __init__(self, parent, nurse_id):
        super().__init__(parent)
        self.title(f"Calendario Turni Infermiere {nurse_id}")
        self.geometry("900x700")
        self.nurse_id = nurse_id
        
        self.current_month = datetime.now().month
        self.current_year = datetime.now().year

        self.state('zoomed')
        self.attributes('-topmost', True)
        
        self.create_widgets()
        self.populate_calendar()

    def create_widgets(self):
        self.header_frame = ctk.CTkFrame(self)
        self.header_frame.pack(fill='x', pady=10)

        self.prev_month_button = ctk.CTkButton(self.header_frame, text="<", command=self.prev_month)
        self.prev_month_button.pack(side='left')

        self.month_label = ctk.CTkLabel(self.header_frame, text="")
        self.month_label.pack(side='left', padx=10)

        self.next_month_button = ctk.CTkButton(self.header_frame, text=">", command=self.next_month)
        self.next_month_button.pack(side='left')

        self.weekday_header_frame = ctk.CTkFrame(self)
        self.weekday_header_frame.pack(fill='x', pady=5)

        self.days_frame = ctk.CTkFrame(self)
        self.days_frame.pack(expand=True, fill='both')

        days_of_week = ["Lun", "Mar", "Mer", "Gio", "Ven", "Sab", "Dom"]
        for i, day in enumerate(days_of_week):
            day_label = ctk.CTkLabel(self.weekday_header_frame, text=day)
            day_label.grid(row=0, column=i, sticky="nsew", padx=5, pady=5)

        # Configurazione delle colonne
        for i in range(7):
            self.weekday_header_frame.grid_columnconfigure(i, weight=1)
            self.days_frame.grid_columnconfigure(i, weight=1)


    def populate_calendar(self):
        shifts = self.retrieve_shifts(self.nurse_id)
        self.display_calendar(shifts)

    def retrieve_shifts(self, nurse_id):
        with connessione_db() as conn:
            cursor = conn.cursor()
            cursor.execute("""
                SELECT c.data,
                    CASE 
                        WHEN c.sala_prelievi = %s THEN 'Sala Prelievi'
                        WHEN c.sala_medicazioni = %s THEN 'Sala Medicazioni'
                        ELSE 'Giorno libero'
                    END AS sala
                FROM calendario c
                WHERE MONTH(data) = %s AND YEAR(data) = %s
                ORDER BY data
            """, (nurse_id, nurse_id, self.current_month, self.current_year))
            shifts = cursor.fetchall()
            return {shift[0]: shift[1] for shift in shifts}

    def display_calendar(self, shifts):
        days_in_month = calendar.monthrange(self.current_year, self.current_month)[1]
        first_day_of_month = calendar.monthrange(self.current_year, self.current_month)[0]

        for widget in self.days_frame.winfo_children():
            widget.destroy()

        self.month_label.configure(text=f"{calendar.month_name[self.current_month]} {self.current_year}")

        date = datetime(self.current_year, self.current_month, 1)
        for i in range(days_in_month):
            current_date = date + timedelta(days=i)
            weekday = current_date.weekday()
            sala = shifts.get(current_date.date(), 'Giorno libero')

            day_label_text = f"{current_date.day}\n{sala}"
            bg_color = "#D0F0C0" if weekday < 5 else "#D3D3D3"

            day_label = ctk.CTkLabel(self.days_frame, text=day_label_text, wraplength=100, bg_color=bg_color)
            day_label.grid(row=(i + first_day_of_month) // 7 + 1, column=weekday, sticky="nsew", padx=5, pady=5)

        for i in range(7):
            self.days_frame.grid_columnconfigure(i, weight=1)

        # Assicurati che le righe siano configurate correttamente
        for j in range((days_in_month + first_day_of_month) // 7 + 1):
            self.days_frame.grid_rowconfigure(j + 1, weight=1)

    def prev_month(self):
        self.current_month -= 1
        if self.current_month == 0:
            self.current_month = 12
            self.current_year -= 1
        self.populate_calendar()

    def next_month(self):
        self.current_month += 1
        if self.current_month == 13:
            self.current_month = 1
            self.current_year += 1
        self.populate_calendar()

class InserisciEsitoInfDialog(ctk.CTkToplevel):
    def __init__(self, parent, infermiere_id=None, infermiere_nome_completo=None):
        super().__init__(parent)
        self.infermiere_id = infermiere_id
        self.infermiere_nome_completo = infermiere_nome_completo
        self.title("Inserisci Esito")
        self.geometry("900x530+300+100")

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        if not self.infermiere_id:
            ctk.CTkLabel(self, text="Seleziona Infermiere").grid(row=0, column=0, pady=5, padx=5, sticky='w')
            self.infermiere_var = ctk.StringVar(self)
            self.infermiere_dict = self.get_infermieri()
            self.infermiere_menu = ctk.CTkOptionMenu(self, values=list(self.infermiere_dict.keys()), variable=self.infermiere_var)
            self.infermiere_menu.grid(row=0, column=1, pady=5, padx=5, sticky='ew')
        ctk.CTkLabel(self, text="Seleziona Paziente").grid(row=0, column=2, pady=5, padx=5, sticky='w')
        self.paziente_var = ctk.StringVar(self)
        self.lista_pazienti = self.get_pazienti()

        # Creazione del CTkOptionMenu con colori personalizzati
        self.paziente_menu = ctk.CTkOptionMenu(
            self, 
            values=list(self.lista_pazienti.keys()), 
            variable=self.paziente_var,
            text_color = "black",
            fg_color="lightgray",  # Colore di sfondo del menu
            button_color="orange",  # Colore del pulsante
            button_hover_color="orange",  # Colore del pulsante al passaggio del mouse
            dropdown_fg_color="white",  # Colore di sfondo delle opzioni
            dropdown_hover_color="lightgray"  # Colore delle opzioni al passaggio del mouse
        )
        self.paziente_menu.grid(row=0, column=3, pady=5, padx=5, sticky='ew')

        
        # Aggiunta di un campo per la data dell'esito
        ctk.CTkLabel(self, text="Data Esito").grid(row=1, column=0, pady=5, padx=5, sticky='w')
        self.data_esito_entry = ctk.CTkEntry(self)
        self.data_esito_entry.insert(0, datetime.now().strftime("%Y-%m-%d"))
        self.data_esito_entry.grid(row=1, column=1, pady=5, padx=5, sticky='w')

        # Aggiunta di un campo per le note
        ctk.CTkLabel(self, text="Note").grid(row=2, column = 0, columnspan = 4,pady=5, padx=5)
        self.note_var = ctk.StringVar(self)
        self.note_text = ctk.CTkTextbox(self, height=330, width = 890)
        self.note_text.grid(row=3, column=0, pady=5, padx=5, columnspan=4, sticky='ew')

        # Puoi legare il contenuto del Textbox alla variabile note_var
        self.note_text.bind("<KeyRelease>", lambda event: self.note_var.set(self.note_text.get("1.0", "end-1c")))

        # Bottone per confermare l'inserimento dell'esito
        ctk.CTkButton(self, text="Invia", command=self.inserisci_esito).grid(row=4, column=0, pady=5, padx=5, columnspan=4)

    def get_infermieri(self):
        
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT ID, Nome, Cognome FROM infermieri")
        infermieri = c.fetchall()
        conn.close()
        return {f"{nome} {cognome}": id for id, nome, cognome in infermieri}

    def get_pazienti(self):
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT ID, Nome, Cognome FROM pazienti")
        pazienti = c.fetchall()
        conn.close()
        return {f"{nome} {cognome}": id for id, nome, cognome in pazienti}

    def inserisci_esito(self):

        if not self.infermiere_id:
            infermiere_selezionato = self.infermiere_var.get()
            if not infermiere_selezionato:
                messagebox.showerror("Errore", "Seleziona un infermiere", parent = self)
                return

            infermiere_id = self.infermiere_dict[infermiere_selezionato]
        else:
            infermiere_id = self.infermiere_id

        paziente_selezionato = self.paziente_var.get()
        if not paziente_selezionato:
            messagebox.showerror("Errore", "Seleziona un paziente", parent = self)
            return

        paziente_id = self.lista_pazienti[paziente_selezionato]

        # Ottieni la data e l'ora corrente
        data_corrente = self.data_esito_entry.get()

        try:
            # Assicurati che la data sia valida convertendola in un oggetto datetime
            data_validata = datetime.strptime(data_corrente, "%Y-%m-%d")
        except ValueError:
            messagebox.showerror("Errore", "Formato data non valido. Usa YYYY-MM-DD.", parent=self)
            return

        # Controlla che ci sia scritto qualcosa nel referto
        esito_messaggio = self.note_var.get()
        if not esito_messaggio:
            messagebox.showerror("Errore", "Compila il referto.", parent = self)
            return

        # Definisci valori fissi per ambulatorio, urgenza, e tipo di visita
        ambulatorio = "Infermieristico"
        urgenza = 0
        tipo_visita = "Infermieristica"

        # Inserisci l'esito nel database con i nuovi valori
        self.inserisci_esito_nel_database(infermiere_id, paziente_id, esito_messaggio, data_validata, ambulatorio, urgenza, tipo_visita)

        messagebox.showinfo("Successo", "Esito inserito con successo", parent=self)
        self.destroy()

    def inserisci_esito_nel_database(self, infermiere_id, paziente_id, esito, data, ambulatorio, urgenza, tipo_visita):
        conn = connessione_db()
        c = conn.cursor()
        c.execute("""
            INSERT INTO esiti (infermiere_id, paziente_id, esito, data, ambulatorio, urgenza, tipo_visita)
            VALUES (%s, %s, %s, %s, %s, %s, %s)
        """, (infermiere_id, paziente_id, esito, data, ambulatorio, urgenza, tipo_visita))
        conn.commit()
        conn.close()
## fine infermiere

# segreteria
class SegreteriaInterface(ctk.CTkToplevel):
    def __init__(self, parent, segreteria,email, password):
        super().__init__(parent)
        self.title("Interfaccia Segreteria")
        self.geometry("1200x600+70+50")
        
        self.segreteria_id = segreteria.get('id')
        self.email = email
        self.password = password

        self.conn = connessione_db()

        self.create_widgets()

    def create_widgets(self):

        self.clear_frame()

        # Connessione al database per ottenere nome e cognome della segreteria
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT nome, cognome FROM segreteria WHERE email = %s AND password = %s", (self.email, self.password))
        segreteria_data = c.fetchone()
        conn.close()

        nome, cognome = segreteria_data

        # Frame superiore per nome e cognome 
        frame_top = ctk.CTkFrame(self, fg_color="#59B7D8", height=50)
        frame_top.pack(fill="x")

        # Nome e cognome 
        nome_label = ctk.CTkLabel(frame_top, text=f"{nome} {cognome}", font=("Long Island", 20))
        nome_label.grid(row=0, column=0, padx=(10, 5), pady=10, sticky="w")

        # Creazione del bottone per il logout
        logout_button = ctk.CTkButton(frame_top, text="Logout", command=self.logout)
        logout_button.grid(row=0, column=1, padx=10, pady=10, sticky="e")

        # Allinea tutto in una griglia con lo spazio rimanente
        frame_top.grid_columnconfigure(0, weight=1)

        # Frame inferiore per i bottoni
        frame_bottom = ctk.CTkFrame(self, height = 50, fg_color="#59B7D8")
        frame_bottom.pack(side="bottom", fill="x")

        # Frame sinistro per il notebook principale
        frame_left = ctk.CTkFrame(self, width = 450, fg_color="white")
        frame_left.pack(side="left", fill="both", expand=True)
        frame_left.pack_propagate(False)

        # Creazione del Tabview principale con larghezza fissa per i tab
        notebook_main = ctk.CTkTabview(frame_left, width=200)
        notebook_main.pack(fill="both", expand=True)

        # Tab per "Pazienti"
        tab_pazienti = notebook_main.add("Pazienti")

        # Tab "Pazienti"
        frame_bottoni_pazienti = ctk.CTkFrame(tab_pazienti)
        frame_bottoni_pazienti.pack(side="bottom", pady=10)

        ctk.CTkButton(frame_bottoni_pazienti, text="Aggiungi Paziente", command=self.nuovo_paziente, fg_color="orange", text_color="black").grid(row=0, column=0, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_pazienti, text="Modifica Paziente", command=lambda: self.apri_modifica_dialog("pazienti"), fg_color="orange", text_color="black").grid(row=0, column=1, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_pazienti, text="Elimina Paziente", command=lambda: self.apri_elimina_dialog("pazienti"), fg_color="orange", text_color="black").grid(row=0, column=3, padx=10, pady=10)
        
        # Configura il frame per distribuire equamente lo spazio tra le colonne
        frame_bottoni_pazienti.grid_columnconfigure(0, weight=1)
        frame_bottoni_pazienti.grid_columnconfigure(1, weight=1)
        frame_bottoni_pazienti.grid_columnconfigure(2, weight=1)

        # Crea e configura la Listbox
        self.listbox_pazienti = tk.Listbox(tab_pazienti, height= 50)
        self.listbox_pazienti.pack(side='top', fill='x')
        self.populate_tutti_pazienti()

        # Tab per "Medici"
        tab_medici = notebook_main.add("Medici")

        # Tab "Medici"
        frame_bottoni_medici = ctk.CTkFrame(tab_medici)
        frame_bottoni_medici.pack(side="bottom", pady=10, expand=True, fill='x')

        ctk.CTkButton(frame_bottoni_medici, text="Nuovo Medico", command=self.inserisci_nuovo_medico, fg_color="orange", text_color="black").grid(row=0, column=0, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_medici, text="Modifica Medico", command=lambda: self.apri_modifica_dialog("medici"), fg_color="orange", text_color="black").grid(row=0, column=1, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_medici, text="Aggiungi Assenza", command=self.open_assenza_dialog, fg_color="orange", text_color="black").grid(row=1, column=0, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_medici, text="Elimina Assenza", command=self.open_elimina_assenza_dialog, fg_color="orange", text_color="black").grid(row=1, column=1, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_medici, text="Elimina Medico", command=lambda: self.apri_elimina_dialog("medici"), fg_color="orange", text_color="black").grid(row=0, column=2, padx=10, pady=10)

        # Configura il frame per distribuire equamente lo spazio tra le colonne
        frame_bottoni_medici.grid_columnconfigure(0, weight=1)
        frame_bottoni_medici.grid_columnconfigure(1, weight=1)
        frame_bottoni_medici.grid_columnconfigure(2, weight=1)

        # Crea e configura la Listbox
        self.listbox_medici = tk.Listbox(tab_medici, height= 50)
        self.listbox_medici.pack(side='top', fill='x')
        self.populate_tutti_medici()

        # Tab per "Infermieri"
        tab_infermieri = notebook_main.add("Infermieri")

        # Tab "Infermieri"
        frame_bottoni_infermieri = ctk.CTkFrame(tab_infermieri)
        frame_bottoni_infermieri.pack(side="bottom", pady=10)

        ctk.CTkButton(frame_bottoni_infermieri, text="Nuovo Infermiere", command=self.inserisci_nuovo_infermiere, fg_color="orange", text_color="black").grid(row=0, column=0, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_infermieri, text="Modifica Infermiere", command=lambda: self.apri_modifica_dialog("infermieri"), fg_color="orange", text_color="black").grid(row=0, column=1, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_infermieri, text="Elimina Infermiere", command=lambda: self.apri_elimina_dialog("infermieri"), fg_color="orange", text_color="black").grid(row=1, column=0, padx=10, pady=10)

        # Crea e configura la Listbox
        self.listbox_infermieri = tk.Listbox(tab_infermieri, height= 50)
        self.listbox_infermieri.pack(side='top', fill='x')
        self.populate_tutti_infermieri()

        # Tab per "Segreteria"
        tab_segreteria = notebook_main.add("Segreteria")

        # Tab "Segreteria"
        frame_bottoni_segreteria = ctk.CTkFrame(tab_segreteria)
        frame_bottoni_segreteria.pack(side="bottom", pady=10)

        ctk.CTkButton(frame_bottoni_segreteria, text="Nuovo Personale", command=self.inserisci_nuovo_personale, fg_color="orange", text_color="black").grid(row=0, column=0, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_segreteria, text="Modifica Personale", command=lambda: self.apri_modifica_dialog("segreteria"), fg_color="orange", text_color="black").grid(row=0, column=1, padx=10, pady=10)
        ctk.CTkButton(frame_bottoni_segreteria, text="Elimina Segreteria", command=lambda: self.apri_elimina_dialog("segreteria"), fg_color="orange", text_color="black").grid(row=1, column=0, padx=10, pady=10)

        # Crea e configura la Listbox
        self.listbox_segreteria = tk.Listbox(tab_segreteria, height= 50)
        self.listbox_segreteria.pack(side='top', fill='x')
        self.populate_tutti_segreteria()

        # Frame centrale
        self.frame_center = ctk.CTkFrame(self, width=350, fg_color="white")
        self.frame_center.pack(side="left", fill="both", expand=True)
        self.frame_center.pack_propagate(False)

        # Frame destro per "Comunicazioni"
        frame_right = ctk.CTkFrame(self,  width=350, fg_color="white")
        frame_right.pack(side="right", fill="both", expand=True)
        frame_right.pack_propagate(False)
        ctk.CTkLabel(frame_right, text="Pazienti - prenotazioni", font=("Long Island", 16)).pack(pady=10)

        # Creazione di un CTkTabview per le comunicazioni (invece di ttk.Notebook)
        notebook_comunicazioni = ctk.CTkTabview(frame_right)
        notebook_comunicazioni.pack(fill="both", expand=True)

        # Tab per "Mie Comunicazioni"
        tab_mie_comunicazioni = notebook_comunicazioni.add("Prenotazioni passate")
        self.textbox_mie_comunicazioni = ctk.CTkTextbox(tab_mie_comunicazioni, width=200, height=400, wrap="word")
        self.textbox_mie_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_mie_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        # Tab per "Tutte le Comunicazioni"
        tab_tutte_comunicazioni = notebook_comunicazioni.add("Prenotazioni future")
        self.textbox_tutte_comunicazioni = ctk.CTkTextbox(tab_tutte_comunicazioni, width=200, height=400, wrap="word")
        self.textbox_tutte_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutte_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        ctk.CTkButton(frame_bottom, text="Inserisci Esito Ambulatoriale", command=self.esito_ambulatoriale).grid(row = 0, column=0, padx = 10, pady= 10)
        ctk.CTkButton(frame_bottom, text="Inserisci Esito Infermiristico", command=self.esito_infermiristico).grid(row = 0, column=1, padx = 10, pady= 10)
        ctk.CTkButton(frame_bottom, text="Storico", command=self.show_history).grid(row = 0, column=3, padx = 10, pady= 10)
        ctk.CTkButton(frame_bottom, text="Aggiorna", command= self.aggiorna).grid(row = 0, column=4, padx = 10, pady= 10)

        frame_bottom.grid_columnconfigure(0, weight=1)
        frame_bottom.grid_columnconfigure(1, weight=1)
        frame_bottom.grid_columnconfigure(2, weight=1)  
        frame_bottom.grid_columnconfigure(3, weight=1)
        
        # Aggiornare le comunicazioni all'avvio
        self.aggiorna_comunicazioni()

    def show_history(self, id = None):
        self.clear_frame()

        # Utilizzo di CTkTextbox al posto di Listbox
        self.text_box = ctk.CTkTextbox(self, height=10, wrap="word", state="normal", yscrollcommand=True)
        self.text_box.pack(pady=20, fill="both", expand=True)

        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        conn = connessione_db()
        cursor = conn.cursor()

        if id: 
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                WHERE p.id = %s
                ORDER BY e.data DESC
            """, (id,))
        else:
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                ORDER BY e.data DESC
            """)
        rows = cursor.fetchall()
        conn.close()

        self.text_box.delete('1.0', tk.END)  # Pulisce il textbox prima di inserire nuovi dati
        if not rows:  # Controlla se la lista è vuota
            self.text_box.insert(tk.END, "Nessun esito registrato")  # Mostra messaggio se non ci sono dati
        else:
            for row in rows:
                urgenza_status = "No" if row[2] == 0 else "Yes"
                healthcare_provider = f"Medico: {row[5]} {row[6]}" if row[5] and row[6] else f"Infermiere: {row[7]} {row[8]}"
                formatted_row = (f"- {row[0]} \t Ambulatorio: {row[1]} \t Urgenza: {urgenza_status} \t "
                                f"Regime della visita: {row[3]} \n "
                                f"{healthcare_provider} \n "
                                f"Paziente: {row[9]} {row[10]}\n Esito: {row[4]} \n\n")
                self.text_box.insert(tk.END, formatted_row) 

    def show_history_esiti(self, id = None, type = None):
        self.clear_frame()

        # Utilizzo di CTkTextbox al posto di Listbox
        self.text_box = ctk.CTkTextbox(self, height=10, wrap="word", state="normal", yscrollcommand=True)
        self.text_box.pack(pady=20, fill="both", expand=True)

        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        conn = connessione_db()
        cursor = conn.cursor()

        if type == "medici": 
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                WHERE m.id = %s
                ORDER BY e.data DESC
            """, (id,))
        elif type == "infermieri":
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                WHERE i.id = %s
                ORDER BY e.data DESC
            """, (id,))

        else:
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                ORDER BY e.data DESC
            """)
        rows = cursor.fetchall()
        conn.close()

        self.text_box.delete('1.0', tk.END)  # Pulisce il textbox prima di inserire nuovi dati
        if not rows:  # Controlla se la lista è vuota
            self.text_box.insert(tk.END, "Nessun esito registrato")  # Mostra messaggio se non ci sono dati
        else:
            for row in rows:
                urgenza_status = "No" if row[2] == 0 else "Yes"
                if type == "medici":
                    formatted_row = (f"- {row[0]} \t Ambulatorio: {row[1]} \t Urgenza: {urgenza_status} \t "
                                    f"Regime della visita: {row[3]} \n "
                                    f"Paziente: {row[9]} {row[10]}\n Esito: {row[4]} \n\n")
                else: 
                    formatted_row = (f"- {row[0]} \t Ambulatorio: {row[1]} \t Urgenza: {urgenza_status} \t "
                                    f"Regime della visita: {row[3]} \n "
                                    f"Paziente: {row[9]} {row[10]}\n Esito: {row[4]} \n\n")
                self.text_box.insert(tk.END, formatted_row) 

    def clear_frame(self):
        for widget in self.winfo_children():
            widget.destroy()

    def esito_infermiristico(self):
        InserisciEsitoInfDialog(self)

    def esito_ambulatoriale(self):
        InserisciEsitoDialog(self, medico_id=None)

    def open_elimina_assenza_dialog(self):
        EliminaAssenzaDialog(self)

    def visualizza_responsabilita(self, id):
        self.clear_frame()  # Pulisce i widget esistenti

        # Bottone per tornare alla vista principale
        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.create_widgets)
        self.back_button.pack(side="bottom", pady=20)

        frame_responsabilita = ctk.CTkFrame(self, width=350)
        frame_responsabilita.pack(side="left", fill="both", expand=True)
        frame_responsabilita.pack_propagate(False)

        # Connessione al database MySQL
        conn = connessione_db()
        c = conn.cursor()  # Usare il cursore che ritorna dizionari

        # Query per recuperare i dettagli del paziente maggiorenne e il conteggio dei minorenni a carico
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico,
                    (SELECT COUNT(*) FROM responsabilita WHERE maggiorenne_id = p.id) AS count_minorenni
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id
                WHERE p.id = %s
            """, (id,))
        paziente = c.fetchone()

        # Query per ottenere i dettagli di ogni minorenne a carico
        c.execute("""
                SELECT p.*
                FROM pazienti p
                JOIN responsabilita r ON p.id = r.minorenne_id
                WHERE r.maggiorenne_id = %s
            """, (id,))
        minorenni = c.fetchall()

        # Mostra i dettagli del maggiorenne
        if paziente:
            ctk.CTkLabel(frame_responsabilita, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(frame_responsabilita, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=5)

            # Mostra i dettagli di ogni minorenne
            for minorenne in minorenni:
                min_frame = ctk.CTkFrame(self, width=350)
                min_frame.pack(side = "left",  fill="both", expand=True)
                min_frame.pack_propagate(False)
                ctk.CTkLabel(min_frame, text=f"Minorenne - Nome: {minorenne[1]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Cognome: {minorenne[2]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"CF: {minorenne[3]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Data di nascita: {minorenne[5]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Luogo di nascita: {minorenne[6]}").pack(pady=5)
                ctk.CTkButton(min_frame, text="Storico", command=lambda min_id=minorenne[0]: self.show_history(min_id)).pack(pady=5)

        else:
            ctk.CTkLabel(frame_responsabilita, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

        conn.close()

    def nuovo_paziente(self):
        NuovoPaziente(self)

    def apri_modifica_dialog(self, tipo):
        ModificaDatiDialog(self, tipo)

    def apri_elimina_dialog(self, tipo):
        EliminaUtenteDialog(self, tipo)

    def aggiorna(self):
        self.populate_tutti_pazienti()
        self.populate_tutti_medici()
        self.populate_tutti_infermieri()
        self.populate_tutti_segreteria()
        self.populate_esiti(self.tab_visite)
        self.aggiorna_comunicazioni()

    def populate_tutti_pazienti(self):

        self.listbox_pazienti.delete(0, tk.END)
    
        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM pazienti ORDER BY cognome")
        pazienti = c.fetchall()
        conn.close()

        for paziente in pazienti:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_pazienti.insert(tk.END, f"{paziente[0]} {paziente[2]} {paziente[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_pazienti.bind('<<ListboxSelect>>', self.seleziona_utente)

    def populate_tutti_medici(self):

        self.listbox_medici.delete(0, tk.END)

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM medici ORDER BY cognome")
        medici = c.fetchall()
        conn.close()

        for medico in medici:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_medici.insert(tk.END, f"{medico[0]} {medico[2]} {medico[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_medici.bind('<<ListboxSelect>>', self.seleziona_medico)

    def populate_tutti_infermieri(self):

        self.listbox_infermieri.delete(0, tk.END)

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM infermieri ORDER BY cognome")
        infermieri = c.fetchall()
        conn.close()

        for infermiere in infermieri:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_infermieri.insert(tk.END, f"{infermiere[0]} {infermiere[2]} {infermiere[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_infermieri.bind('<<ListboxSelect>>', self.seleziona_infermiere)

    def populate_tutti_segreteria(self):

        self.listbox_segreteria.delete(0, tk.END)

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT id, nome, cognome FROM segreteria ORDER BY cognome")
        segreteria = c.fetchall()
        conn.close()

        for personale in segreteria:
            # Aggiunge ogni paziente alla Listbox
            self.listbox_segreteria.insert(tk.END, f"{personale[0]} {personale[2]} {personale[1]}")

            # Configura il comando da eseguire quando un elemento è selezionato
            # Attenzione: è necessario implementare show_patient_details per gestire la selezione
            self.listbox_segreteria.bind('<<ListboxSelect>>', self.seleziona_personale)

    def seleziona_utente(self, event):
        index = self.listbox_pazienti.curselection()
        if index:
            self.id_selezionato = self.listbox_pazienti.get(index[0]).split(" - ")[0]
            self.show_patient_details(self.id_selezionato)

    def seleziona_medico(self, event):
        index = self.listbox_medici.curselection()
        if index:
            self.id_selezionato = self.listbox_medici.get(index[0]).split(" - ")[0]
            self.show_medico_details(self.id_selezionato)

    def seleziona_infermiere(self, event):
        index = self.listbox_infermieri.curselection()
        if index:
            self.id_selezionato = self.listbox_infermieri.get(index[0]).split(" - ")[0]
            self.show_infermiere_details(self.id_selezionato)

    def seleziona_personale(self, event):
        index = self.listbox_segreteria.curselection()
        if index:
            self.id_selezionato = self.listbox_segreteria.get(index[0]).split(" - ")[0]
            self.show_segreteria_details(self.id_selezionato)

    def show_patient_details(self, patient_id):
        for widget in self.frame_center.winfo_children():
            widget.destroy()

        conn = connessione_db()
        # Configura il cursore per restituire dizionari
        c = conn.cursor()
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico ,r.minorenne_id
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id 
                LEFT JOIN responsabilita r ON p.id = r.maggiorenne_id
                WHERE p.id = %s
            """, (patient_id,))
        paziente = c.fetchone()
        conn.close()

        if paziente:
            ctk.CTkLabel(self.frame_center, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(self.frame_center, text=f"Password: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(self.frame_center, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=5)


            # Controlla se il paziente ha minorenni a carico
            if paziente[11] is not None:
                ctk.CTkButton(self.frame_center, text="Visualizza Responsabilità", command=lambda: self.visualizza_responsabilita(paziente[0])).pack(pady=10)
        else:
            ctk.CTkLabel(self.frame_center, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

    def show_medico_details(self, patient_id):
            for widget in self.frame_center.winfo_children():
                widget.destroy()

            conn = connessione_db()
            c = conn.cursor()
            c.execute("SELECT * FROM medici WHERE id = %s", (patient_id,))
            medico = c.fetchone()
            conn.close()

            if medico:
                ctk.CTkLabel(self.frame_center, text=f"Nome: {medico[1]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Cognome: {medico[2]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Email: {medico[4]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Password: {medico[5]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"CF: {medico[6]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Specialità: {medico[7]}").pack(pady=5)
                ctk.CTkButton(self.frame_center, text="Esito", command=lambda: self.show_history_esiti(medico[0], "medici")).pack(pady=5)
            else:
                ctk.CTkLabel(self.frame_center, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

    def show_infermiere_details(self, patient_id):
            for widget in self.frame_center.winfo_children():
                widget.destroy()

            conn = connessione_db()
            c = conn.cursor()
            c.execute("SELECT * FROM infermieri WHERE id = %s", (patient_id,))
            infermiere = c.fetchone()
            conn.close()

            if infermiere:
                ctk.CTkLabel(self.frame_center, text=f"Nome: {infermiere[1]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Cognome: {infermiere[2]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"CF: {infermiere[3]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Email: {infermiere[4]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Password: {infermiere[5]}").pack(pady=5)
                ctk.CTkButton(self.frame_center, text="Esito", command=lambda: self.show_history_esiti(infermiere[0], "infermieri")).pack(pady=5)

            else:
                ctk.CTkLabel(self.frame_center, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

    def show_segreteria_details(self, patient_id):
            for widget in self.frame_center.winfo_children():
                widget.destroy()

            conn = connessione_db()
            c = conn.cursor()
            c.execute("SELECT * FROM segreteria WHERE id = %s", (patient_id,))
            segreteria = c.fetchone()
            conn.close()

            if segreteria:
                ctk.CTkLabel(self.frame_center, text=f"Nome: {segreteria[1]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Cognome: {segreteria[2]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Email: {segreteria[3]}").pack(pady=5)
                ctk.CTkLabel(self.frame_center, text=f"Password: {segreteria[4]}").pack(pady=5)

            else:
                ctk.CTkLabel(self.frame_center, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

    def inserisci_nuovo_medico(self):
        # Funzione per aprire una nuova finestra per inserire i dettagli del nuovo medico
        self.new_window = ctk.CTkToplevel(self)
        self.new_window.title("Inserisci Nuovo Medico")
        self.new_window.attributes('-topmost', True)

        # Creazione delle etichette e dei campi di input
        ctk.CTkLabel(self.new_window, text="Nome").grid(row=0, column=0, padx=10, pady=5)
        self.entry_nome = ctk.CTkEntry(self.new_window)
        self.entry_nome.grid(row=0, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Cognome").grid(row=1, column=0, padx=10, pady=5)
        self.entry_cognome = ctk.CTkEntry(self.new_window)
        self.entry_cognome.grid(row=1, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Codice Fiscale").grid(row=2, column=0, padx=10, pady=5)
        self.entry_cf = ctk.CTkEntry(self.new_window)
        self.entry_cf.grid(row=2, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Email").grid(row=3, column=0, padx=10, pady=5)
        self.entry_email = ctk.CTkEntry(self.new_window)
        self.entry_email.grid(row=3, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Password").grid(row=4, column=0, padx=10, pady=5)
        self.entry_password = ctk.CTkEntry(self.new_window, show="*")
        self.entry_password.grid(row=4, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Specialità (opzionale)").grid(row=5, column=0, padx=10, pady=5)
        self.entry_specialita = ctk.CTkEntry(self.new_window)
        self.entry_specialita.grid(row=5, column=1, padx=10, pady=5)

        # Bottone per inserire il nuovo medico
        btn_inserisci = ctk.CTkButton(self.new_window, text="Inserisci", command=self.inserisci_medico_database)
        btn_inserisci.grid(row=6, column=0, columnspan=2, padx=10, pady=10)

    def inserisci_medico_database(self):
        # Funzione per inserire il nuovo medico nel database
        nome = self.entry_nome.get().strip()
        cognome = self.entry_cognome.get().strip()
        codice_fiscale = self.entry_cf.get().strip()
        email = self.entry_email.get().strip()
        password = self.entry_password.get().strip()
        specialita = self.entry_specialita.get().strip()

        # Funzione di controllo per verificare se il campo contiene solo lettere
        def is_valid_text(text):
            return text.isalpha()

        # Controllo per verificare se tutti i campi obbligatori sono compilati
        if not (nome and cognome and codice_fiscale and email and password):
            messagebox.showwarning("Attenzione", "Per favore, compila tutti i campi richiesti", parent=self.new_window)
            return

        # Controlli di validità sui campi di testo
        if not is_valid_text(nome):
            messagebox.showwarning("Attenzione", "Il campo 'Nome' deve contenere solo lettere", parent=self.new_window)
            return

        if not is_valid_text(cognome):
            messagebox.showwarning("Attenzione", "Il campo 'Cognome' deve contenere solo lettere", parent=self.new_window)
            return

        if not codice_fiscale.isalnum() or len(codice_fiscale) < 16:
            messagebox.showwarning("Attenzione", "Il Codice Fiscale deve essere di 16 caratteri alfanumerici", parent=self.new_window)
            return

        if '@' not in email or '.' not in email:
            messagebox.showwarning("Attenzione", "Inserisci un'email valida", parent=self.new_window)
            return

        if len(password) < 8:
            messagebox.showwarning("Attenzione", "La password deve contenere almeno 8 caratteri", parent=self.new_window)
            return
        
        if specialita and not is_valid_text(specialita):
            messagebox.showwarning("Attenzione", "Il campo 'Specialità' deve contenere solo lettere se compilato", parent=self.new_window)
            return

        try:
            # Connessione al database MySQL
            conn = connessione_db()
            c = conn.cursor()

            # Verifica se il medico esiste già
            c.execute("SELECT id FROM medici WHERE CF = %s OR email = %s", (codice_fiscale, email))
            if c.fetchone():
                messagebox.showerror("Errore", "Un medico con lo stesso codice fiscale o email esiste già.", parent=self.new_window)
                return

            # Inserimento dei dati nel database
            c.execute("INSERT INTO medici (nome, cognome, CF, email, password, specialita) VALUES (%s, %s, %s, %s, %s, %s)",
                    (nome, cognome, codice_fiscale, email, password, specialita))
            conn.commit()

            messagebox.showinfo("Successo", "Nuovo medico inserito correttamente nel database.", parent=self.new_window)
        except mysql.connector.Error as err:
            messagebox.showerror("Errore", f"Errore del database: {err}", parent=self.new_window)
        finally:
            if conn.is_connected():
                c.close()
                conn.close()
                self.new_window.destroy()

    def inserisci_nuovo_infermiere(self):
        # Funzione per aprire una nuova finestra per inserire i dettagli del nuovo infermiere
        self.new_window = ctk.CTkToplevel(self)
        self.new_window.title("Inserisci Nuovo Infermiere")
        self.new_window.attributes('-topmost', True)

        ctk.CTkLabel(self.new_window, text="Nome").grid(row=0, column=0, padx=10, pady=5)
        self.entry_nome = ctk.CTkEntry(self.new_window)
        self.entry_nome.grid(row=0, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Cognome").grid(row=1, column=0, padx=10, pady=5)
        self.entry_cognome = ctk.CTkEntry(self.new_window)
        self.entry_cognome.grid(row=1, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Codice Fiscale").grid(row=2, column=0, padx=10, pady=5)
        self.entry_cf = ctk.CTkEntry(self.new_window)
        self.entry_cf.grid(row=2, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Email").grid(row=3, column=0, padx=10, pady=5)
        self.entry_email = ctk.CTkEntry(self.new_window)
        self.entry_email.grid(row=3, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Password").grid(row=4, column=0, padx=10, pady=5)
        self.entry_password = ctk.CTkEntry(self.new_window, show="*")
        self.entry_password.grid(row=4, column=1, padx=10, pady=5)

        btn_inserisci = ctk.CTkButton(self.new_window, text="Inserisci", command=self.inserisci_infermiere_database)
        btn_inserisci.grid(row=5, column=0, columnspan=2, padx=10, pady=10)

    def inserisci_infermiere_database(self):
        nome = self.entry_nome.get().strip()
        cognome = self.entry_cognome.get().strip()
        codice_fiscale = self.entry_cf.get().strip()
        email = self.entry_email.get().strip()
        password = self.entry_password.get().strip()

        # Funzione di controllo per verificare se il campo contiene solo lettere
        def is_valid_text(text):
            return text.isalpha()

        # Controllo per verificare se tutti i campi obbligatori sono compilati
        if not (nome and cognome and codice_fiscale and email and password):
            messagebox.showwarning("Attenzione", "Per favore, compila tutti i campi richiesti", parent=self.new_window)
            return

        # Controlli di validità sui campi di testo
        if not is_valid_text(nome):
            messagebox.showwarning("Attenzione", "Il campo 'Nome' deve contenere solo lettere", parent=self.new_window)
            return

        if not is_valid_text(cognome):
            messagebox.showwarning("Attenzione", "Il campo 'Cognome' deve contenere solo lettere", parent=self.new_window)
            return

        if not codice_fiscale.isalnum() or len(codice_fiscale) != 16:
            messagebox.showwarning("Attenzione", "Il Codice Fiscale deve essere di 16 caratteri alfanumerici", parent=self.new_window)
            return

        if '@' not in email or '.' not in email:
            messagebox.showwarning("Attenzione", "Inserisci un'email valida", parent=self.new_window)
            return

        if len(password) < 8:
            messagebox.showwarning("Attenzione", "La password deve contenere almeno 8 caratteri", parent=self.new_window)
            return

        try:
            conn = connessione_db()
            c = conn.cursor()
            c.execute("SELECT id FROM infermieri WHERE CF = %s OR email = %s", (codice_fiscale, email))
            if c.fetchone():
                messagebox.showerror("Errore", "Un infermiere con lo stesso codice fiscale o email esiste già.", parent=self.new_window)
                return

            c.execute("INSERT INTO infermieri (nome, cognome, CF, email, password) VALUES (%s, %s, %s, %s, %s)",
                    (nome, cognome, codice_fiscale, email, password))
            conn.commit()

            messagebox.showinfo("Successo", "Nuovo infermiere inserito correttamente nel database.", parent=self.new_window)
        except mysql.connector.Error as err:
            messagebox.showerror("Errore", f"Errore del database: {err}", parent=self.new_window)
        finally:
            if conn.is_connected():
                c.close()
                conn.close()
                self.new_window.destroy()

    def inserisci_nuovo_personale(self):
        # Funzione per aprire una nuova finestra per inserire i dettagli del nuovo personale segreteria
        self.new_window = ctk.CTkToplevel(self)
        self.new_window.attributes('-topmost', True)
        self.new_window.title("Inserisci Nuovo Personale")

        ctk.CTkLabel(self.new_window, text="Nome").grid(row=0, column=0, padx=10, pady=5)
        self.entry_nome = ctk.CTkEntry(self.new_window)
        self.entry_nome.grid(row=0, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Cognome").grid(row=1, column=0, padx=10, pady=5)
        self.entry_cognome = ctk.CTkEntry(self.new_window)
        self.entry_cognome.grid(row=1, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Email").grid(row=2, column=0, padx=10, pady=5)
        self.entry_email = ctk.CTkEntry(self.new_window)
        self.entry_email.grid(row=2, column=1, padx=10, pady=5)

        ctk.CTkLabel(self.new_window, text="Password").grid(row=3, column=0, padx=10, pady=5)
        self.entry_password = ctk.CTkEntry(self.new_window, show="*")
        self.entry_password.grid(row=3, column=1, padx=10, pady=5)

        btn_inserisci = ctk.CTkButton(self.new_window, text="Inserisci", command=self.inserisci_personale_database)
        btn_inserisci.grid(row=4, column=0, columnspan=2, padx=10, pady=10)
    
    def inserisci_personale_database(self):
        # Funzione per inserire il nuovo infermiere nel database
        nome = self.entry_nome.get()
        cognome = self.entry_cognome.get()
        email = self.entry_email.get()
        password = self.entry_password.get()

        # Funzione di controllo per verificare se il campo contiene solo lettere
        def is_valid_text(text):
            return text.isalpha()

        # Controllo per verificare se tutti i campi obbligatori sono compilati
        if not (nome and cognome and email and password):
            messagebox.showwarning("Attenzione", "Per favore, compila tutti i campi richiesti", parent=self.new_window)
            return

        # Controlli di validità sui campi di testo
        if not is_valid_text(nome):
            messagebox.showwarning("Attenzione", "Il campo 'Nome' deve contenere solo lettere", parent=self.new_window)
            return

        if not is_valid_text(cognome):
            messagebox.showwarning("Attenzione", "Il campo 'Cognome' deve contenere solo lettere", parent=self.new_window)
            return

        if '@' not in email or '.' not in email:
            messagebox.showwarning("Attenzione", "Inserisci un'email valida", parent=self.new_window)
            return

        if len(password) < 8:
            messagebox.showwarning("Attenzione", "La password deve contenere almeno 8 caratteri", parent=self.new_window)
            return

        try:
            # Connessione al database MySQL
            conn = connessione_db()
            c = conn.cursor()

            # Verifica se il medico esiste già
            c.execute("SELECT id FROM segreteria WHERE email = %s", (email,))
            if c.fetchone():
                messagebox.showerror("Errore", "Un personale della segreteria con lo stesso codice fiscale o email esiste già.", parent=self.new_window)
                return

            # Inserimento dei dati nel database
            c.execute("INSERT INTO segreteria (nome, cognome, email, password) VALUES (%s, %s,  %s, %s)",
                    (nome, cognome, email, password))
            conn.commit()

            messagebox.showinfo("Successo", "Nuovo personale inserito correttamente nel database.", parent=self.new_window)
        except mysql.connector.Error as err:
            messagebox.showerror("Errore", f"Errore del database: {err}", parent=self.new_window)
        finally:
            if conn.is_connected():
                c.close()
                conn.close()
                self.new_window.destroy()

    def open_assenza_dialog(self):
        AggiungiAssenzaDialog(self)

    def aggiorna_comunicazioni(self):

        c = self.conn.cursor()

        # Ottenere la data odierna
        oggi = datetime.now().date()

        # Recuperare le comunicazioni della segreteria
        c.execute("""
            SELECT p.medico_id, p.data, p.sala, p.orario, p.tipo_visita, 
                CONCAT(pa.nome, ' ', pa.cognome) AS nome_completo_paziente,
                CONCAT(m.nome, ' ', m.cognome) AS nome_completo_medico, 
                CONCAT(i.nome, ' ', i.cognome) AS nome_completo_infermiere
            FROM prenotazioni p
            LEFT JOIN pazienti pa ON p.paziente_id = pa.id
            LEFT JOIN medici m ON p.medico_id = m.id
            LEFT JOIN infermieri i ON p.infermiere_id = i.id
            WHERE p.data < %s
            ORDER BY p.data DESC
            """, (oggi,))
        mie_comunicazioni = c.fetchall()

        # Recuperare tutte le comunicazioni
        c.execute("""
            SELECT p.medico_id, p.data, p.sala, p.orario, p.tipo_visita, 
                CONCAT(pa.nome, ' ', pa.cognome) AS nome_completo_paziente,
                CONCAT(m.nome, ' ', m.cognome) AS nome_completo_medico, 
                CONCAT(i.nome, ' ', i.cognome) AS nome_completo_infermiere
            FROM prenotazioni p
            LEFT JOIN pazienti pa ON p.paziente_id = pa.id
            LEFT JOIN medici m ON p.medico_id = m.id
            LEFT JOIN infermieri i ON p.infermiere_id = i.id
            WHERE p.data >= %s
            ORDER BY p.data ASC
            """, (oggi,))
        tutte_comunicazioni = c.fetchall()
        
        c.close()

        if mie_comunicazioni:
            mie_comunicazioni_str = "\n".join(
                [
                    f"- Il {data} alle {orario} in {sala}. \n"
                    f"Prenotazione per paziente {nome_completo_paziente} \n"
                    f"{f'con medico {nome_completo_medico}' if medico_id else f'con infermiere {nome_completo_infermiere}'} \n"
                    f"Tipo di visita: {tipo_visita}"
                    for medico_id, data, sala, orario, tipo_visita,
                    nome_completo_paziente, nome_completo_medico, nome_completo_infermiere in mie_comunicazioni
                ]
            )
        else:
            mie_comunicazioni_str = "Nessuna prenotazione antecedente trovata."

        if tutte_comunicazioni:
            tutte_comunicazioni_str = "\n".join(
                [
                    f"- Il {data} alle {orario} in {sala}. \n"
                    f"Prenotazione per paziente {nome_completo_paziente} \n"
                    f"{f'con medico {nome_completo_medico}' if medico_id else f'con infermiere {nome_completo_infermiere}'} \n"
                    f"Tipo di visita: {tipo_visita}"
                    for medico_id, data, sala, orario, tipo_visita,
                    nome_completo_paziente, nome_completo_medico, nome_completo_infermiere in tutte_comunicazioni
                ]
            )
        else:
            tutte_comunicazioni_str = "Nessuna prenotazione trovata."

        # Aggiornare il contenuto delle Textbox
        self.textbox_mie_comunicazioni.configure(state="normal")
        self.textbox_mie_comunicazioni.delete("1.0", "end")
        self.textbox_mie_comunicazioni.insert("1.0", mie_comunicazioni_str)
        self.textbox_mie_comunicazioni.configure(state="disabled")

        self.textbox_tutte_comunicazioni.configure(state="normal")
        self.textbox_tutte_comunicazioni.delete("1.0", "end")
        self.textbox_tutte_comunicazioni.insert("1.0", tutte_comunicazioni_str)
        self.textbox_tutte_comunicazioni.configure(state="disabled")

    def logout(self):
        # Mostra una finestra di dialogo di conferma
        risposta = messagebox.askyesno("Conferma Logout", "Sei sicuro di voler uscire?", parent = self)
        
        if risposta:  # Se l'utente conferma
            # Chiudi la connessione al database se è aperta
            if self.conn is not None and self.conn.is_connected():
                self.conn.close()
            
            # Chiudi la finestra corrente
            self.destroy()

class NuovoPaziente(ctk.CTkToplevel):
    def __init__(self, parent):
        super().__init__(parent)
        self.title("Inserisci nuovo paziente")
        self.geometry("500x550")
        self.transient(parent)
        
        self.create_widgets()
    
    def create_widgets(self):

        # Etichette e campi di input per i dati del paziente
        labels_texts = ["Nome:", "Cognome:", "Email:", "Password:", "Codice Sanitario:", "Medico curante:", "Data di nascita (YYYY-MM-DD):", "Luogo di nascita:", "Ha meno di 14 anni?", "Se sì, seleziona il responsabile:"]
        self.entries = []

        for row, text in enumerate(labels_texts):
            ctk.CTkLabel(self, text=text).grid(row=row, column=0, padx=10, pady=5, sticky="e")

        self.entry_nome = ctk.CTkEntry(self)
        self.entry_nome.grid(row=0, column=1, padx=10, pady=5)

        self.entry_cognome = ctk.CTkEntry(self)
        self.entry_cognome.grid(row=1, column=1, padx=10, pady=5)

        self.entry_email = ctk.CTkEntry(self)
        self.entry_email.grid(row=2, column=1, padx=10, pady=5)

        self.entry_password = ctk.CTkEntry(self)
        self.entry_password.grid(row=3, column=1, padx=10, pady=5)

        self.entry_codice = ctk.CTkEntry(self)
        self.entry_codice.grid(row=4, column=1, padx=10, pady=5)

        # Medico curante
        medici = self.ottieni_elenco_medici()
        self.medici_dict = {nome: id for id, nome in medici}
        self.medico_curante = ctk.StringVar(self)
        ctk.CTkOptionMenu(self, values=list(self.medici_dict.keys()), variable=self.medico_curante).grid(row=5, column=1, padx=10, pady=5)

        self.entry_data_nascita = ctk.CTkEntry(self)
        self.entry_data_nascita.grid(row=6, column=1, padx=10, pady=5)

        self.entry_luogo_nascita = ctk.CTkEntry(self)
        self.entry_luogo_nascita.grid(row=7, column=1, padx=10, pady=5)

        # È minorenne?
        self.responsabile = ctk.BooleanVar()
        ctk.CTkCheckBox(self, text="Sì", variable=self.responsabile, command=self.update_email_based_on_responsabile).grid(row=8, column=1, padx=10, pady=5, sticky="w")

        # Inizializzazione del widget di selezione del responsabile:
        self.maggiorenne = ctk.StringVar(self)
        responsabili = self.ottieni_elenco_maggiorenni()  # Assumi che ritorni un dizionario {id: nome}
        self.maggiorenni_dict = {nome: id for id, nome in responsabili}
        responsabili_option_menu = ctk.CTkOptionMenu(self, values=list(self.maggiorenni_dict.keys()), variable=self.maggiorenne)
        responsabili_option_menu.grid(row=9, column=1, padx=10, pady=5)

        # Tracciamento delle modifiche nella selezione del responsabile:
        self.maggiorenne.trace_add('write', lambda *args: self.update_email_based_on_responsabile())

        # Bottone per salvare i dati del paziente
        ctk.CTkButton(self, text="Salva", command=self.salva_paziente).grid(row=10, column=0, columnspan=2, pady=20)

    def ottieni_elenco_medici(self):
        # Funzione per ottenere l'elenco dei medici dal database
        conn = connessione_db()
        c = conn.cursor()
        
        c.execute("SELECT ID, Nome, Cognome FROM medici")
        medici_rows = c.fetchall()
        conn.close()
        
        medici = [(row[0], f"{row[1]} {row[2]}") for row in medici_rows]
        return medici

    def ottieni_elenco_maggiorenni(self):
        # Funzione per ottenere l'elenco dei pazienti maggiorenni dal database
        conn = connessione_db()
        c = conn.cursor()
        
        c.execute("SELECT ID, Nome, Cognome FROM pazienti WHERE TIMESTAMPDIFF(YEAR, Nascita, CURDATE()) >= 18")
        maggiorenni_rows = c.fetchall()
        conn.close()
        
        maggiorenni = [(row[0], f"{row[1]} {row[2]}") for row in maggiorenni_rows]
        return maggiorenni

    def paziente_esiste(self, nome, cognome, codice, data_nascita_str, luogo_nascita, email):
        # Funzione per verificare se un paziente esiste già nel database
        
        # Connessione al database MySQL
        conn = connessione_db()
        c = conn.cursor()
        
        # Esecuzione della query per verificare se il paziente esiste
        c.execute("SELECT * FROM pazienti \
                  WHERE Nome = %s \
                  AND Cognome = %s \
                  AND CS = %s \
                  AND Nascita = %s\
                  AND Luogo_Nascita = %s\
                  AND Email= %s", (nome, cognome, codice, data_nascita_str, luogo_nascita, email))
        result = c.fetchone()
        
        # Chiusura della connessione al database
        conn.close()
        
        # Restituzione del risultato della query (True se il paziente esiste già, False altrimenti)
        return bool(result)

    def update_email_based_on_responsabile(self):
        if self.responsabile.get():  # Se è selezionato "È minorenne? Sì"
            responsabile_id = self.maggiorenni_dict.get(self.maggiorenne.get())
            if responsabile_id:
                responsabile_email = self.get_email_by_responsabile_id(responsabile_id)
                self.entry_email.delete(0, 'end')
                self.entry_email.insert(0, responsabile_email)

    def get_email_by_responsabile_id(self, responsabile_id):
        # Connessione al database
        conn = connessione_db()
        cursor = conn.cursor()

        # Esegue la query per ottenere l'email basata sull'ID del responsabile
        query = "SELECT email FROM pazienti WHERE id = %s"
        cursor.execute(query, (responsabile_id,))
        result = cursor.fetchone()
        conn.close()  # Chiudi la connessione dopo l'uso

        # Restituisce l'email se trovata, altrimenti restituisce una stringa vuota
        return result[0] if result else ""

    def salva_paziente(self):
        # Recupero dei valori inseriti dall'utente
        nome = self.entry_nome.get()
        cognome = self.entry_cognome.get()
        email = self.entry_email.get()
        password = self.entry_password.get()
        codice = self.entry_codice.get()
        medico_curante_nome = self.medico_curante.get()
        medico_curante_id = self.medici_dict[medico_curante_nome]
        data_nascita_str = self.entry_data_nascita.get()
        luogo_nascita = self.entry_luogo_nascita.get()
        responsabilita = self.responsabile.get()
        maggiorenne_nome = self.maggiorenne.get()
        maggiorenne_id = self.maggiorenni_dict[maggiorenne_nome] if responsabilita else None

        # Funzione di controllo per verificare se il campo contiene solo lettere
        def is_valid_text(text):
            return text.isalpha()

        # Verifica se tutti i campi obbligatori sono stati inseriti
        if not nome or not cognome or not data_nascita_str or not luogo_nascita or not email or not password:
            messagebox.showerror("Errore", "Per favore, inserisci tutti i campi.", parent = self)
            return
        
        # Controllo se il paziente esiste già nel database
        if self.paziente_esiste(nome, cognome, codice, data_nascita_str, luogo_nascita, email):
            messagebox.showerror("Errore", "Il paziente esiste già nel database!", parent = self)
            return

        # Controlli di validità sui campi di testo
        if not is_valid_text(nome):
            messagebox.showwarning("Attenzione", "Il campo 'Nome' deve contenere solo lettere", parent=self)
            return

        if not is_valid_text(cognome):
            messagebox.showwarning("Attenzione", "Il campo 'Cognome' deve contenere solo lettere", parent=self)
            return

        if not codice.isalnum() or len(codice) < 16:
            messagebox.showwarning("Attenzione", "Il Codice Fiscale deve essere di 16 caratteri alfanumerici", parent=self)
            return

        if '@' not in email or '.' not in email:
            messagebox.showwarning("Attenzione", "Inserisci un'email valida", parent=self)
            return

        if len(password) < 8:
            messagebox.showwarning("Attenzione", "La password deve contenere almeno 8 caratteri", parent=self)
            return
        
        data_nascita = datetime.strptime(data_nascita_str, '%Y-%m-%d').date()
        data_odierna = datetime.now().date()

        # Confronta la data inserita con la data corrente
        if data_nascita > data_odierna:
            messagebox.showerror("Errore", "La data di nascita non può essere futura.", parent=self)
            return
        
        # Calcola l'età per determinare se il paziente è minorenne
        age = data_odierna.year - data_nascita.year - ((data_odierna.month, data_odierna.day) < (data_nascita.month, data_nascita.day))
        if age < 14:
            # Verifica che un responsabile sia stato selezionato se il paziente è minorenne
            if not self.maggiorenne.get() or self.maggiorenne.get() == "Seleziona un responsabile":
                messagebox.showerror("Errore", "Per un paziente che ha meno di 14 anni è obbligatorio selezionare un responsabile.", parent=self)
                return

        # Connessione al database MySQL
        conn = connessione_db()
        c = conn.cursor()
        
        # Inserimento dei dati nel database
        try:
            # Conversione della stringa della data in un oggetto datetime
            data_nascita = datetime.strptime(data_nascita_str, "%Y-%m-%d").date()
            # Inserimento dei dati nel database
            c.execute("INSERT INTO pazienti (Nome, Cognome, CS, Medico_Curante, Nascita, Luogo_Nascita, Email, Password) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)",
                    (nome, cognome, codice, medico_curante_id, data_nascita, luogo_nascita, email, password))
            paziente_id = c.lastrowid
            if responsabilita and maggiorenne_id:
                c.execute("INSERT INTO responsabilita (maggiorenne_id, minorenne_id) VALUES (%s, %s)", (maggiorenne_id, paziente_id))
            conn.commit()
            # Messaggio di conferma
            messagebox.showinfo("Successo", "Paziente aggiunto con successo al database!", parent = self)
            # Chiusura della finestra di dialogo
            self.destroy()
        except Exception as e:
            # Gestione dell'errore in caso di problemi durante l'inserimento
            messagebox.showerror("Errore", f"Si è verificato un errore durante l'inserimento del paziente: {str(e)}", parent = self)
            # Rollback delle modifiche nel database
            conn.rollback()
        finally:
            # Chiusura della connessione al database
            conn.close()

class EliminaUtenteDialog(ctk.CTkToplevel):
    def __init__(self, parent, tipo):
        super().__init__(parent)
        self.tipo = tipo
        self.title(f"Elimina {tipo.capitalize()}")
        self.geometry("400x400")

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale
        self.id_selezionato = None

        ctk.CTkLabel(self, text=f"Seleziona un {tipo} dalla lista:").pack(pady=5)

        self.listbox_utenti = tk.Listbox(self, width=100)
        self.listbox_utenti.pack(pady=5)
        self.listbox_utenti.bind('<<ListboxSelect>>', self.seleziona_utente)

        self.button_conferma_eliminazione = ctk.CTkButton(self, text="Conferma Eliminazione", command=self.conferma_eliminazione)
        self.button_conferma_eliminazione.pack(pady=20)

        self.ottieni_utenti()

    def ottieni_utenti(self):
        conn = connessione_db()
        c = conn.cursor()
        c.execute(f"SELECT ID, Nome, Cognome FROM {self.tipo}")
        utenti = c.fetchall()
        conn.close()

        for utente in utenti:
            utente_str = " - ".join(map(str, utente))
            self.listbox_utenti.insert(ctk.END, utente_str)

    def seleziona_utente(self, event):
        index = self.listbox_utenti.curselection()
        if index:
            self.id_selezionato = self.listbox_utenti.get(index).split(" - ")[0]
        else:
            messagebox.showerror("Errore", f"Per favore, seleziona un {self.tipo} dalla lista.", parent = self)

    def conferma_eliminazione(self):
        if not self.id_selezionato:
            messagebox.showerror("Errore", f"Seleziona un {self.tipo} dalla lista.", parent = self)
            return

        conferma = messagebox.askyesno("Conferma Eliminazione", f"Sei sicuro di voler eliminare questo {self.tipo}?", parent = self)
        if conferma:
            conn = connessione_db()
            c = conn.cursor()
            try:
                if self.tipo == "pazienti":
                    # Eliminare tutte le righe collegate nella tabella comunicazioni_paziente
                    c.execute("DELETE FROM esiti WHERE paziente_id = %s", (self.id_selezionato,))
                    c.execute("DELETE FROM responsabilita WHERE maggiorenne_id = %s", (self.id_selezionato,))
                    # Eliminare il paziente
                    c.execute("DELETE FROM pazienti WHERE ID = %s", (self.id_selezionato,))
                elif self.tipo == "medici":
                    # Eliminare tutte le righe collegate nella tabella comunicazioni_medico
                    c.execute("DELETE FROM esiti WHERE medico_id = %s", (self.id_selezionato,))
                    # Aggiungi qui il codice per eliminare le righe collegate nelle altre tabelle
                    c.execute("DELETE FROM medici WHERE ID = %s", (self.id_selezionato,))
                elif self.tipo == "infermieri":
                    # Eliminare tutte le righe collegate nella tabella comunicazioni_infermiere
                    c.execute("DELETE FROM esiti WHERE infermiere_id = %s", (self.id_selezionato,))
                    # Aggiungi qui il codice per eliminare le righe collegate nelle altre tabelle
                    c.execute("DELETE FROM infermieri WHERE ID = %s", (self.id_selezionato,))
                elif self.tipo == "segreteria":
                    c.execute("DELETE FROM esiti WHERE segreteria_id = %s", (self.id_selezionato,))
                    # Aggiungi qui il codice per eliminare le righe collegate nelle altre tabelle
                    c.execute("DELETE FROM segreteria WHERE ID = %s", (self.id_selezionato,))
                conn.commit()
                messagebox.showinfo("Successo", f"{self.tipo.capitalize()} eliminato con successo!",parent = self)
                self.destroy()
            except Exception as e:
                messagebox.showerror("Errore", f"Si è verificato un errore durante l'eliminazione: {str(e)}", parent = self)
                conn.rollback()
            finally:
                conn.close()
## fine segreteria

# paziente
class PazienteInterface(ctk.CTkToplevel):
    def __init__(self, parent, paziente, email, password):
        super().__init__(parent)
        self.title("Interfaccia Paziente")
        self.geometry("1200x600+70+50")

        self.paziente = paziente.get('id')
        self.email = email
        self.password = password
        self.conn = connessione_db()

        self.crea_interfaccia()
    
    def crea_interfaccia(self):

        self.clear_frame()

        conn = connessione_db()
        c = conn.cursor()
        c.execute("SELECT nome, cognome FROM pazienti WHERE email = %s AND password = %s", (self.email, self.password))
        segreteria_data = c.fetchone()
        conn.close()

        nome, cognome = segreteria_data

        # Frame superiore per nome e cognome 
        frame_top = ctk.CTkFrame(self, fg_color="#23F094", height=50)
        frame_top.pack(fill="x", side="top")

        # Nome e cognome 
        nome_label = ctk.CTkLabel(frame_top, text=f"{nome} {cognome}",  font=("Long Island", 20))
        nome_label.pack(side="left", padx=(10, 5), pady=10)

        # Creazione del bottone per il logout
        logout_button = ctk.CTkButton(frame_top, text="Logout", command=self.logout)
        logout_button.pack(side="right", padx=10, pady=10)

        # Allinea tutto in una griglia con lo spazio rimanente
        frame_top.grid_columnconfigure(0, weight=1)

        # Frame inferiore per i bottoni
        frame_bottom = ctk.CTkFrame(self, height = 50, fg_color="#23F094")
        frame_bottom.pack(side="bottom", fill="x")

        # Frame sinistro per i dati del paziente
        frame_sinistro = ctk.CTkFrame(self, fg_color="white")
        frame_sinistro.pack(side="left", fill="y")

        self.labels_dati_paziente = {}
        for campo in ["Nome", "Cognome", "Codice Sanitario", "Medico Curante", "Data di Nascita", "Luogo di Nascita"]:
            campo_frame = ctk.CTkFrame(frame_sinistro, fg_color="white")
            campo_frame.pack(fill="x", pady=2)
            
            label_titolo = ctk.CTkLabel(campo_frame, text=campo + ": ", anchor="w", width=15 , fg_color="white")
            label_titolo.pack(side="left", padx=5)
            
            label_valore = ctk.CTkLabel(campo_frame, text="", fg_color="white")
            label_valore.pack(side="left")
            
            self.labels_dati_paziente[campo] = label_valore

        self.aggiorna_dati_paziente()

        button_modifica_dati = ctk.CTkButton(frame_sinistro, text="Modifica Dati", command=lambda: self.modifica_dati(self.paziente))
        button_modifica_dati.pack(side="bottom", pady=10)

        # bottone per modificare i dati del minorenne
        conn = connessione_db()
        cursor = conn.cursor()
        cursor.execute("SELECT COUNT(*) FROM responsabilita WHERE maggiorenne_id = %s", (self.paziente,))
        result = cursor.fetchone()
        conn.close()

        if result[0] > 0:  # Se ci sono minorenni a carico
            self.button_modifica_minorenne = ctk.CTkButton(frame_sinistro, text="Modifica Minorenne", command=self.apri_modifica_minorenni)
            self.button_modifica_minorenne.pack(side="bottom", pady=10)

            self.button_storico_minorenne = ctk.CTkButton(frame_sinistro, text = "Storico", command=lambda: self.visualizza_responsabilita(self.paziente))
            self.button_storico_minorenne.pack(side="bottom",pady= 10)

        # Creazione delle tabview al centro
        frame_centrale = ctk.CTkFrame(self, width=450, fg_color="white")
        frame_centrale.pack(side="left", fill="both", expand=True)
        frame_centrale.pack_propagate(False)

        self.tab_control = ctk.CTkTabview(frame_centrale)
        self.tab_control.pack(fill="both", expand=True)

        # Creazione delle schede all'interno della Tabview e delle CTkTextbox
        self.tab_visite_prestazioni = self.tab_control.add("Visite e Prestazioni")
        self.text_visite_prestazioni = ctk.CTkTextbox(self.tab_visite_prestazioni, height=10, wrap='word')
        self.text_visite_prestazioni.pack(fill="both", expand=True)

        self.tab_visite = self.tab_control.add("Visite")
        self.text_visite = ctk.CTkTextbox(self.tab_visite, height=10, wrap='word')
        self.text_visite.pack(fill="both", expand=True)

        self.tab_prestazioni = self.tab_control.add("Prestazioni")
        self.text_prestazioni = ctk.CTkTextbox(self.tab_prestazioni, height=10, wrap='word')
        self.text_prestazioni.pack(fill="both", expand=True)

        # Frame destro per "Prenotazioni"
        frame_right = ctk.CTkFrame(self, width=450, fg_color="white")
        frame_right.pack(side="right", fill="both")
        frame_right.pack_propagate(False)
        ctk.CTkLabel(frame_right, text="Prenotazioni", font=("Helvetica", 14)).pack(pady=10)

        button_aggiorna = ctk.CTkButton(frame_right, text="Aggiorna", command= self.aggiorna_comunicazioni_paziente)
        button_aggiorna.pack(side="bottom", padx=5, pady=5)

        # Creazione del Tabview per "Prenotazioni"
        notebook_comunicazioni = ctk.CTkTabview(frame_right)
        notebook_comunicazioni.pack(fill="both", expand=True)

        # Tab per "Comunicazioni Interne"
        tab_mie_comunicazioni = notebook_comunicazioni.add("Comunicazioni interne")
        self.textbox_mie_comunicazioni = ctk.CTkTextbox(tab_mie_comunicazioni, width=200, height=350, wrap="word")
        self.textbox_mie_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_mie_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        # Tab per "Prossime prenotazioni"
        tab_tutte_comunicazioni = notebook_comunicazioni.add("Prossime prenotazioni")
        self.textbox_tutte_comunicazioni = ctk.CTkTextbox(tab_tutte_comunicazioni, width=200, height=350, wrap="word")
        self.textbox_tutte_comunicazioni.pack(fill="both", expand=True, padx=10, pady=10)
        self.textbox_tutte_comunicazioni.configure(state="disabled")  # Rende il textbox non modificabile dall'utente

        button_prenota_visita = ctk.CTkButton(frame_bottom, text="Prenota Visita Medico", command=lambda: self.prenota_visita_medico(self.paziente))
        button_prenota_visita.pack(side="left", padx=5, pady=5)

        button_prenota_prestazione = ctk.CTkButton(frame_bottom, text="Prenota Visita Infermiere", command=lambda: self.prenota_visita_infermiere(self.paziente))
        button_prenota_prestazione.pack(side="right", padx=5, pady=5)


        self.popola_tutte_le_tab()
        self.aggiorna_comunicazioni_paziente()

    def visualizza_responsabilita(self, id):
        self.clear_frame()  # Pulisce i widget esistenti

        # Bottone per tornare alla vista principale
        self.back_button = ctk.CTkButton(self, text="Indietro", command=self.crea_interfaccia)
        self.back_button.pack(side="bottom", pady=20)

        frame_responsabilita = ctk.CTkFrame(self, width=350)
        frame_responsabilita.pack(side="left", fill="both", expand=True)
        frame_responsabilita.pack_propagate(False)

        # Connessione al database MySQL
        conn = connessione_db()
        c = conn.cursor()  # Usare il cursore che ritorna dizionari

        # Query per recuperare i dettagli del paziente maggiorenne e il conteggio dei minorenni a carico
        c.execute("""
                SELECT p.*, m.nome AS nome_medico, m.cognome AS cognome_medico,
                    (SELECT COUNT(*) FROM responsabilita WHERE maggiorenne_id = p.id) AS count_minorenni
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id
                WHERE p.id = %s
            """, (id,))
        paziente = c.fetchone()

        # Query per ottenere i dettagli di ogni minorenne a carico
        c.execute("""
                SELECT p.*
                FROM pazienti p
                JOIN responsabilita r ON p.id = r.minorenne_id
                WHERE r.maggiorenne_id = %s
            """, (id,))
        minorenni = c.fetchall()

        # Mostra i dettagli del maggiorenne
        if paziente:
            ctk.CTkLabel(frame_responsabilita, text=f"Nome: {paziente[1]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Cognome: {paziente[2]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"CF: {paziente[3]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Medico curante: {paziente[9]} {paziente[10]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Data di nascita: {paziente[5]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Luogo di nascita: {paziente[6]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[7]}").pack(pady=5)
            ctk.CTkLabel(frame_responsabilita, text=f"Email: {paziente[8]}").pack(pady=5)
            ctk.CTkButton(frame_responsabilita, text="Storico", command=lambda: self.show_history(paziente[0])).pack(pady=5)

            # Mostra i dettagli di ogni minorenne
            for minorenne in minorenni:
                min_frame = ctk.CTkFrame(self, width=350)
                min_frame.pack(side = "left",  fill="both", expand=True)
                min_frame.pack_propagate(False)
                ctk.CTkLabel(min_frame, text=f"Minorenne - Nome: {minorenne[1]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Cognome: {minorenne[2]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"CF: {minorenne[3]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Data di nascita: {minorenne[5]}").pack(pady=5)
                ctk.CTkLabel(min_frame, text=f"Luogo di nascita: {minorenne[6]}").pack(pady=5)
                ctk.CTkButton(min_frame, text="Storico", command=lambda min_id=minorenne[0]: self.show_history(min_id)).pack(pady=5)

        else:
            ctk.CTkLabel(frame_responsabilita, text="Nessun dettaglio disponibile per questo paziente.").pack(pady=20)

        conn.close()

    def show_history(self, id = None):
        self.clear_frame()

        # Utilizzo di CTkTextbox al posto di Listbox
        self.text_box = ctk.CTkTextbox(self, height=10, wrap="word", state="normal", yscrollcommand=True)
        self.text_box.pack(pady=20, fill="both", expand=True)

        self.back_button = ctk.CTkButton(self, text="Indietro", command=lambda: self.visualizza_responsabilita(self.paziente))
        self.back_button.pack(side="bottom", pady=20)

        conn = connessione_db()
        cursor = conn.cursor()

        if id: 
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                WHERE p.id = %s
                ORDER BY e.data DESC
            """, (id,))
        else:
            cursor.execute("""
                SELECT e.data, e.ambulatorio, e.urgenza, e.tipo_visita, e.esito, 
                    m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                    i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere,
                    p.Nome AS NomePaziente, p.Cognome AS CognomePaziente
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                LEFT JOIN infermieri i ON e.infermiere_id = i.ID
                LEFT JOIN pazienti p ON e.paziente_id = p.ID
                ORDER BY e.data DESC
            """)
        rows = cursor.fetchall()
        conn.close()

        self.text_box.delete('1.0', tk.END)  # Pulisce il textbox prima di inserire nuovi dati
        if not rows:  # Controlla se la lista è vuota
            self.text_box.insert(tk.END, "Nessun esito registrato")  # Mostra messaggio se non ci sono dati
        else:
            for row in rows:
                urgenza_status = "No" if row[2] == 0 else "Yes"
                healthcare_provider = f"Medico: {row[5]} {row[6]}" if row[5] and row[6] else f"Infermiere: {row[7]} {row[8]}"
                formatted_row = (f"- {row[0]} \t Ambulatorio: {row[1]} \t Urgenza: {urgenza_status} \t "
                                f"Regime della visita: {row[3]} \n "
                                f"{healthcare_provider} \n "
                                f"Paziente: {row[9]} {row[10]}\n Esito: {row[4]}\n\n")
                self.text_box.insert(tk.END, formatted_row) 

    def apri_modifica_minorenni(self):
        finestra_minorenni = ctk.CTkToplevel(self)
        finestra_minorenni.title("Modifica Minorenni")
        finestra_minorenni.geometry("400x300")

        finestra_minorenni.attributes('-topmost', True)

        ctk.CTkLabel(finestra_minorenni, text="Seleziona un minorenne dalla lista:").pack(pady=5)

        # Crea una Listbox per mostrare i minorenni
        self.listbox_utenti = tk.Listbox(finestra_minorenni, width=50)
        self.listbox_utenti.pack(pady=5)

        # Bind per selezionare l'utente
        self.listbox_utenti.bind('<<ListboxSelect>>', self.seleziona_utente)

        # Bottone per modificare l'utente
        #self.button_modifica = ctk.CTkButton(finestra_minorenni, text="Modifica", command=lambda: self.modifica_dati())
        #self.button_modifica.pack(pady=20)

        # Recupera e mostra i minorenni associati al paziente
        self.ottieni_utenti()

    def ottieni_utenti(self):
        # Connessione al database per ottenere i minorenni associati al maggiorenne
        conn = connessione_db()
        c = conn.cursor()

        # Esegui la query per ottenere i minorenni associati al maggiorenne
        c.execute("""
            SELECT p.id, p.Nome, p.Cognome
            FROM pazienti p
            JOIN responsabilita r ON p.id = r.minorenne_id
            WHERE r.maggiorenne_id = %s
        """, (self.paziente,))

        # Recupera i risultati e chiudi la connessione
        utenti = c.fetchall()
        conn.close()

        # Aggiunge ogni minorenne alla Listbox
        for utente in utenti:
            self.listbox_utenti.insert(tk.END, f"{utente[0]} - {utente[1]} {utente[2]}")

    def seleziona_utente(self, event):
        index = self.listbox_utenti.curselection()
        if index:
            self.id_selezionato = self.listbox_utenti.get(index).split(" - ")[0]
            ModificaDatiPazienteDialog(self, self.id_selezionato)
        else:
            messagebox.showerror("Errore", "Per favore, seleziona un minorenne dalla lista.", parent=self)

    def popola_tutte_le_tab(self):
        # Connessione al database
        conn = connessione_db()
        c = conn.cursor()

        # Query per ottenere tutte le visite e prestazioni
        query_visite_prestazioni = """
            SELECT * FROM ((SELECT 
                e.data, e.tipo_visita, e.esito, 
                CONCAT(m.Nome, ' ', m.Cognome) AS medico, 
                CONCAT(i.Nome, ' ', i.Cognome) AS infermiere
            FROM esiti e
            LEFT JOIN medici m ON e.medico_id = m.ID
            LEFT JOIN infermieri i ON e.infermiere_id = i.ID
            WHERE e.paziente_id = %s)
            UNION
            (
            SELECT 
                e.data, e.tipo_visita, e.esito, 
                CONCAT(m.Nome, ' ', m.Cognome) AS medico, 
                CONCAT(i.Nome, ' ', i.Cognome) AS infermiere
            FROM esiti e
            LEFT JOIN medici m ON e.medico_id = m.ID
            LEFT JOIN infermieri i ON e.infermiere_id = i.ID
            JOIN responsabilita r ON e.paziente_id = r.minorenne_id
            WHERE r.maggiorenne_id = %s)
            ) AS combined_results
            ORDER BY combined_results.data DESC
        """
        c.execute(query_visite_prestazioni, (self.paziente, self.paziente))
        esiti = c.fetchall()

        # Query per ottenere solo le visite
        query_visite = """
            SELECT * FROM (
                (SELECT 
                    e.data, e.tipo_visita, e.esito, 
                    CONCAT(m.Nome, ' ', m.Cognome) AS medico
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                WHERE e.paziente_id = %s AND e.tipo_visita IS NOT NULL AND e.medico_id IS NOT NULL)
                UNION
                (SELECT 
                    e.data, e.tipo_visita, e.esito, 
                    CONCAT(m.Nome, ' ', m.Cognome) AS medico
                FROM esiti e
                LEFT JOIN medici m ON e.medico_id = m.ID
                JOIN responsabilita r ON e.paziente_id = r.minorenne_id
                WHERE r.maggiorenne_id = %s AND e.tipo_visita IS NOT NULL AND e.medico_id IS NOT NULL)
            ) AS combined_results
            ORDER BY combined_results.data DESC
        """
        c.execute(query_visite, (self.paziente, self.paziente))

        visite = c.fetchall()

        # Query per ottenere solo le prestazioni
        query_prestazioni = """
            SELECT 
                e.data, e.esito, 
                CONCAT(i.Nome, ' ', i.Cognome) AS infermiere
            FROM esiti e
            LEFT JOIN infermieri i ON e.infermiere_id = i.ID
            WHERE e.paziente_id = %s AND e.infermiere_id IS NOT NULL
            ORDER BY e.data DESC
        """
        c.execute(query_prestazioni, (self.paziente,))
        prestazioni = c.fetchall()

        # Chiudere la connessione al database
        conn.close()

        esiti_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')}\n" +
            (f"Medico: {medico}" if medico else "") +  # Aggiunge dettagli del medico se presenti
            (f"Infermiere: {infermiere}" if infermiere else "") +  # Aggiunge dettagli dell'infermiere se presenti
            f"\n{tipo_visita} \nEsito: {esito}\n"
            for data, tipo_visita, esito, medico, infermiere in esiti]
        ) if esiti else "Nessuna prenotazione registrata."

        visite_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')}\nMedico: {medico} / {tipo_visita}\nEsito: {esito}\n"
            for data, tipo_visita, esito, medico in visite]
        ) if visite else "Nessuna prenotazione."

        prestazioni_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')}\nInfermiere: {infermiere}\nEsito: {esito}\n"
            for data, esito, infermiere in prestazioni]
        ) if prestazioni else "Nessuna prenotazione."


        # Aggiornare il contenuto delle Textbox
        self.text_visite_prestazioni.configure(state="normal")
        self.text_visite_prestazioni.delete("1.0", "end")
        self.text_visite_prestazioni.insert("1.0", esiti_str)
        self.text_visite_prestazioni.configure(state="disabled")

        self.text_visite.configure(state="normal")
        self.text_visite.delete("1.0", "end")
        self.text_visite.insert("1.0", visite_str)
        self.text_visite.configure(state="disabled")

        # Aggiornare il contenuto delle Textbox
        self.text_prestazioni.configure(state="normal")
        self.text_prestazioni.delete("1.0", "end")
        self.text_prestazioni.insert("1.0", prestazioni_str)
        self.text_prestazioni.configure(state="disabled")

    def aggiorna_dati_paziente(self):
        # Recupera e mostra i dati del paziente
        conn = connessione_db()
        c = conn.cursor()
        c.execute("""
                SELECT p.*, CONCAT(m.nome, ' ', m.cognome) AS nome_completo_medico, r.minorenne_id
                FROM pazienti p
                LEFT JOIN medici m ON p.medico_curante = m.id 
                LEFT JOIN responsabilita r ON p.id = r.maggiorenne_id
                WHERE p.id = %s
            """, (self.paziente,))
        result = c.fetchone()
        conn.close()


        if result:
            dati = {
                "Nome": result[1],
                "Cognome": result[2],
                "Codice Sanitario": result[3],
                "Medico Curante": result [9], 
                "Data di Nascita": result[5],
                "Luogo di Nascita": result[6]
            }
            for campo, valore in dati.items():
                self.labels_dati_paziente[campo].configure(text=valore)
 
    def modifica_dati(self, id):
        ModificaDatiPazienteDialog(self, id)

    def prenota_visita_medico(self, paziente):
        prenota_finestra = ctk.CTkToplevel()  # Creiamo una finestra secondaria
        prenota_finestra.title("Prenota Appuntamento")
        prenota_finestra.geometry("500x500")
        prenota_finestra.attributes('-topmost', True)
        
        # Etichetta e ComboBox per scegliere il medico
        ctk.CTkLabel(prenota_finestra, text="Seleziona Medico").pack(pady=10)
        medico_combobox = ttk.Combobox(prenota_finestra, state="readonly")
        medico_combobox.pack(pady=10)
        
        # Riempire la ComboBox con i medici disponibili
        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)
        
        query = "SELECT id, nome, cognome, specialita FROM medici"
        cursor.execute(query)
        medici = cursor.fetchall()
        
        medico_combobox['values'] = [f"{medico['id']} - {medico['nome']} {medico['cognome']} ({medico['specialita']})" for medico in medici]
        medico_combobox.current(0)  # Seleziona il primo medico per default

        def calculate_mindate():
            today = datetime.now()
            # Aumenta la data se oggi è sabato (5) o domenica (6)
            while today.weekday() > 4:
                today += timedelta(days=1)
            return today

        # Seleziona il giorno con DateEntry
        ctk.CTkLabel(prenota_finestra, text="Seleziona Giorno").pack(pady=10)
        giorno_entry = DateEntry(prenota_finestra, date_pattern='yyyy-mm-dd', mindate=calculate_mindate(), maxdate=datetime.now().replace(day=31, month=12))  # Limitiamo ai giorni disponibili
        giorno_entry.pack(pady=10)
        
        # Seleziona l'orario con una Combobox
        ctk.CTkLabel(prenota_finestra, text="Seleziona Orario").pack(pady=10)
        orari_disponibili = [f"{h:02}:{m:02}" for h in range(9, 18) for m in range(0, 60, 15)]
        orario_combobox = ttk.Combobox(prenota_finestra, state="readonly", values=orari_disponibili)
        orario_combobox.pack(pady=10)
        orario_combobox.current(0)

        # Seleziona il tipo di visita con una Combobox
        ctk.CTkLabel(prenota_finestra, text="Seleziona Tipo di Visita").pack(pady=10)
        tipi_visita = ["Per adulti", "Pediatrica"]
        tipo_visita_combobox = ttk.Combobox(prenota_finestra, state="readonly", values=tipi_visita)
        tipo_visita_combobox.pack(pady=10)
        tipo_visita_combobox.current(0)

        # Aggiungi un bottone per la visita d'urgenza
        ctk.CTkLabel(prenota_finestra, text="Oppure").pack(pady=5)
        urgenza_button = ctk.CTkButton(prenota_finestra, text="Richiedi Visita d'Urgenza", command=lambda: self.prenota_urgenza(paziente))
        urgenza_button.pack(pady=10)

        conn.close()
        
        # Bottone per confermare la prenotazione
        prenota_button = ctk.CTkButton(prenota_finestra, text="Prenota", command=lambda: self.conferma_prenotazione
                                       (paziente, medico_combobox, giorno_entry, orario_combobox, 
                                        tipo_visita_combobox, prenota_finestra))
        prenota_button.pack(pady=20)

        # Aggiorna la listbox con i nuovi appuntamenti
        self.popola_tutte_le_tab()
    
    def conferma_prenotazione(self, paziente_id, medico_combobox, giorno_entry, orario_combobox, tipo_visita_combobox, finestra):
        medico_selezionato = medico_combobox.get()
        medico_id = int(medico_selezionato.split(" - ")[0])  # Estraiamo l'ID del medico
        specialita_medico = medico_selezionato.split(" (")[1][:-1]  # Estraiamo la specialità del medico
        
        # Recuperiamo la data e l'orario scelti
        data_prenotazione = giorno_entry.get_date().strftime('%Y-%m-%d')

        orario_prenotazione = orario_combobox.get()
        orario_attuale = datetime.now().strftime("%H:%M")
        oggi = datetime.now().strftime("%Y-%m-%d")  # Ottieni la data di oggi

        # Verifica se il giorno selezionato è oggi
        if data_prenotazione == oggi:
            if orario_prenotazione < orario_attuale:
                messagebox.showerror("Errore", "Non si può scegliere un orario passato per il giorno di oggi.", parent=self)
                conn.close()
                return
        
        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)

        # Verifica se il medico è assente nel giorno scelto
        query_assenze = """
        SELECT inizio, fine FROM assenze
        WHERE medico = %s AND %s BETWEEN inizio AND fine
        """
        cursor.execute(query_assenze, (medico_id, data_prenotazione))
        assenza = cursor.fetchone()

        if assenza:
            # Cerca un sostituto con la stessa specialità
            query_sostituto = """
            SELECT id, nome, cognome FROM medici
            WHERE specialita = %s AND id != %s AND id NOT IN (
                SELECT medico FROM assenze WHERE %s BETWEEN inizio AND fine
            )
            """
            cursor.execute(query_sostituto, (specialita_medico, medico_id, data_prenotazione))
            sostituto = cursor.fetchone()

            if sostituto:
                # Proponi il sostituto trovato
                risposta = messagebox.askyesno("Medico Assente", 
                                            f"Il medico selezionato è assente dal {assenza['inizio']} al {assenza['fine']}.\n"
                                            f"Vuoi prenotare con il Dr. {sostituto['nome']} {sostituto['cognome']} (Specialità: {specialita_medico})?", parent = self)
                if risposta:
                    medico_id = sostituto['id']
                else:
                    conn.close()
                    return
            else:
                # Cerca un altro medico disponibile (anche con specialità diversa)
                query_altro_medico = """
                SELECT id, nome, cognome, specialita FROM medici
                WHERE id != %s AND id NOT IN (
                    SELECT medico FROM assenze WHERE %s BETWEEN inizio AND fine
                )
                """
                cursor.execute(query_altro_medico, (medico_id, data_prenotazione))
                altro_medico = cursor.fetchone()

                if altro_medico:
                    risposta = messagebox.askyesno("Medico Assente", 
                                                f"Il medico selezionato è assente dal {assenza['inizio']} al {assenza['fine']}.\n"
                                                f"Vuoi prenotare con il Dr. {altro_medico['nome']} {altro_medico['cognome']} (Specialità: {altro_medico['specialita']})?", parent = self)
                    if risposta:
                        medico_id = altro_medico['id']
                    else:
                        conn.close()
                        return
                else:
                    messagebox.showerror("Errore", "Non ci sono medici disponibili in questa data.", parent = self)
                    conn.close()
                    return

        # Verifica se il medico è già impegnato per l'orario e data selezionati
        query_verifica_medico = """
        SELECT * FROM prenotazioni
        WHERE medico_id = %s AND data = %s AND orario = %s
        """
        cursor.execute(query_verifica_medico, (medico_id, data_prenotazione, orario_prenotazione))
        risultato_medico = cursor.fetchone()

        if risultato_medico:
            messagebox.showerror("Errore", "Il medico selezionato è già impegnato in questo orario.", parent = self)
            conn.close()
            return

        # Verifica se il paziente ha già un appuntamento nello stesso giorno e orario
        query_verifica_paziente = """
        SELECT * FROM prenotazioni
        WHERE paziente_id = %s AND data = %s AND orario = %s
        """
        cursor.execute(query_verifica_paziente, (paziente_id, data_prenotazione, orario_prenotazione))
        risultato_paziente = cursor.fetchone()

        if risultato_paziente:
            messagebox.showerror("Errore", "Hai già un appuntamento in questo orario.", parent = self)
            conn.close()
            return

        tipo_visita = tipo_visita_combobox.get() # per adulti / pediatrica

        # Definisci le sale disponibili in base al tipo di visita
        if tipo_visita == "Per adulti":
            sale_disponibili = ["Sala 1", "Sala 2", "Sala 3"]
        elif tipo_visita == "Pediatrica":
            sale_disponibili = ["Pediatrica"]

        sala_assegnata = None

        for sala in sale_disponibili:
            query_verifica_sala = """
            SELECT * FROM prenotazioni
            WHERE sala = %s AND data = %s AND orario = %s
            """
            cursor.execute(query_verifica_sala, (sala, data_prenotazione, orario_prenotazione))
            risultato_sala = cursor.fetchone()

            if not risultato_sala:
                sala_assegnata = sala
                break
        
        if not sala_assegnata:
            messagebox.showerror("Errore", "Orario scelto non disponibile. Tutte le sale sono occupate.", parent = self)
            conn.close()
            return

        # Inserire la prenotazione nel database con la sala assegnata
        risposta = messagebox.askyesno("Conferma Prenotazione", "Vuoi confermare la prenotazione?",  parent = self)

        if risposta:  # Se l'utente clicca 'Sì'
            query_inserisci_prenotazione = """
            INSERT INTO prenotazioni (paziente_id, medico_id, infermiere_id, data, sala, orario, tipo_visita)
            VALUES (%s, %s, NULL, %s, %s, %s, %s)
            """
            cursor.execute(query_inserisci_prenotazione, (paziente_id, medico_id, data_prenotazione, sala_assegnata, orario_prenotazione, tipo_visita))
            conn.commit()
            messagebox.showinfo("Prenotazione Effettuata", f"La prenotazione è stata effettuata con successo! Sala assegnata: {sala_assegnata}", parent = self)
        else:  # Se l'utente clicca 'No'
            messagebox.showinfo("Annullato", "Prenotazione annullata.",  parent = self)
        
        # Chiudiamo la finestra di prenotazione
        finestra.destroy()
        
        conn.close()

        # Aggiorna la listbox con i nuovi appuntamenti
        self.aggiorna_comunicazioni_paziente()

    def prenota_urgenza(self, paziente_id):
        conn = connessione_db()
        cursor = conn.cursor(dictionary=True)

        # Trova il primo medico disponibile
        query_medico_disponibile = """
        SELECT m.id, m.nome, m.cognome, m.specialita
        FROM medici m
        LEFT JOIN assenze a ON m.id = a.medico AND CURDATE() BETWEEN a.inizio AND a.fine
        WHERE a.medico IS NULL
        ORDER BY m.id ASC
        LIMIT 1
        """
        cursor.execute(query_medico_disponibile)
        medico_disponibile = cursor.fetchone()

        if medico_disponibile:
            # Calcola l'orario successivo disponibile arrotondando ai 15 minuti più vicini
            now = datetime.now()
            minuti = (now.minute // 15 + 1) * 15
            orario_urgenza = now.replace(minute=0, second=0, microsecond=0) + timedelta(minutes=minuti)
            
            # Verifica che l'orario sia all'interno delle 09:00 e 18:00
            if orario_urgenza.hour < 9:
                orario_urgenza = orario_urgenza.replace(hour=9, minute=0)
            elif orario_urgenza.hour >= 18:
                orario_urgenza = orario_urgenza.replace(hour=9, minute=0) + timedelta(days=1)
            
            data_urgenza = orario_urgenza.strftime('%Y-%m-%d')
            orario_urgenza_str = orario_urgenza.strftime('%H:%M')

            # Verifica che il paziente non abbia già un appuntamento nello stesso orario
            query_verifica_paziente = """
            SELECT * FROM prenotazioni
            WHERE paziente_id = %s AND data = %s AND orario = %s
            """
            cursor.execute(query_verifica_paziente, (paziente_id, data_urgenza, orario_urgenza_str))
            appuntamento_esistente = cursor.fetchone()

            if appuntamento_esistente:
                messagebox.showerror("Errore", "Hai già un appuntamento in questo orario.")
                conn.close()
                return

            # Verifica la disponibilità della sala
            sale_disponibili = ["Sala 1", "Sala 2", "Sala 3"]
            sala_assegnata = None

            for sala in sale_disponibili:
                query_verifica_sala = """
                SELECT * FROM prenotazioni
                WHERE sala = %s AND data = %s AND orario = %s
                """
                cursor.execute(query_verifica_sala, (sala, data_urgenza, orario_urgenza_str))
                risultato_sala = cursor.fetchone()

                if not risultato_sala:
                    sala_assegnata = sala
                    break

            if not sala_assegnata:
                messagebox.showerror("Errore", "Tutte le sale sono occupate per questo orario.")
                conn.close()
                return

            # Messaggio di conferma per il paziente
            conferma = messagebox.askyesno(
                "Medico Trovato",
                f"Il medico disponibile per la visita d'urgenza è:\n\n"
                f"Medico: {medico_disponibile['nome']} {medico_disponibile['cognome']} ({medico_disponibile['specialita']})\n"
                f"Data: {data_urgenza}\nOrario: {orario_urgenza_str}\nSala: {sala_assegnata}\n\n"
                "Vuoi fissare questo appuntamento?"
            )

            if conferma:
                # Inserisci la prenotazione nel database
                query_inserisci_prenotazione = """
                INSERT INTO prenotazioni (paziente_id, medico_id, infermiere_id, data, sala, orario, tipo_visita)
                VALUES (%s, %s, NULL, %s, %s, %s, 'Urgenza')
                """
                cursor.execute(query_inserisci_prenotazione, (paziente_id, medico_disponibile['id'], data_urgenza, sala_assegnata, orario_urgenza_str))
                conn.commit()

                # Notifica per il paziente
                query_paziente = "SELECT email FROM pazienti WHERE id = %s"
                cursor.execute(query_paziente, (paziente_id,))
                paziente_email = cursor.fetchone()['email']
                messagebox.showinfo("Prenotazione Confermata", f"La prenotazione è stata effettuata con successo! Una notifica è stata inviata a {paziente_email}.")

                # Notifica per il medico
                query_medico = "SELECT email FROM medici WHERE id = %s"
                cursor.execute(query_medico, (medico_disponibile['id'],))
                medico_email = cursor.fetchone()['email']
                messagebox.showinfo("Notifica Medico", f"Una notifica è stata inviata al medico {medico_email} riguardo l'urgenza.")

            else:
                messagebox.showinfo("Prenotazione Annullata", "La prenotazione d'urgenza è stata annullata.")

        else:
            messagebox.showerror("Errore", "Nessun medico disponibile per la visita d'urgenza.")
        
        conn.close()

        # Aggiorna la listbox con i nuovi appuntamenti
        self.popola_tutte_le_tab()

    def prenota_visita_infermiere(self, paziente_id):
        prenota_finestra = ctk.CTkToplevel()
        prenota_finestra.title("Prenota Visita Infermieristica")
        prenota_finestra.geometry("600x400")
        prenota_finestra.attributes('-topmost', True)
        self.paziente_id = paziente_id
        self.infermiere_id = None  # Sarà selezionato al termine

        def calculate_mindate():
            today = datetime.now()
            # Aumenta la data se oggi è sabato (5) o domenica (6)
            while today.weekday() > 4:
                today += timedelta(days=1)
            return today
        
        def update_infermiere_id(*args):
            data_scelta = giorno_entry.get()
            sala_scelta = sala_var.get()
            if data_scelta and sala_scelta:  # Assicurati che sia la data che la sala siano state selezionate
                try:
                    conn = connessione_db()  # Assicurati che questa funzione sia definita per aprire la connessione al DB
                    c = conn.cursor()
                    sala_column = 'sala_prelievi' if sala_scelta == "Sala Prelievi" else 'sala_medicazioni'
                    c.execute(f"SELECT {sala_column} FROM calendario WHERE data = %s", (data_scelta,))
                    result = c.fetchone()
                    if result:
                        self.infermiere_id = result[0]
                    else:
                        messagebox.showerror("Errore", "Nessun infermiere trovato per la data e sala selezionata.")
                    conn.close()
                except Exception as e:
                    messagebox.showerror("Errore Database", str(e))

        # Seleziona il giorno con DateEntry
        ctk.CTkLabel(prenota_finestra, text="Seleziona Giorno").pack(pady=10)
        giorno_entry = DateEntry(prenota_finestra, date_pattern='yyyy-mm-dd', mindate=calculate_mindate(), maxdate=datetime.now().replace(day=31, month=12))  # Limitiamo ai giorni disponibili
        giorno_entry.pack(pady=10)
        
        # Seleziona l'orario con una Combobox
        ctk.CTkLabel(prenota_finestra, text="Seleziona Orario").pack(pady=10)
        orari_disponibili = [f"{h:02}:{m:02}" for h in range(9, 18) for m in range(0, 60, 15)]
        orario_combobox = ttk.Combobox(prenota_finestra, state="readonly", values=orari_disponibili)
        orario_combobox.pack(pady=10)
        orario_combobox.current(0)

        # Sala
        ctk.CTkLabel(prenota_finestra, text="Scegli la sala:").pack(pady=5)
        sala_var = ctk.StringVar(prenota_finestra)
        sala_menu = ctk.CTkOptionMenu(prenota_finestra, values=["Sala Medicazioni", "Sala Prelievi"], variable=sala_var)
        sala_menu.pack(pady=5)

        # Aggiungi i binding agli eventi
        giorno_entry.bind("<<DateEntrySelected>>", update_infermiere_id)
        sala_var.trace('w', update_infermiere_id)

        # Bottone di conferma
        ctk.CTkButton(prenota_finestra, text="Conferma Prenotazione", command=lambda: self.conferma_prenotazione_infermiere(giorno_entry, orario_combobox, sala_var, prenota_finestra)).pack(pady=20)

    def conferma_prenotazione_infermiere(self, giorno_entry, orario_combobox, sala_var, prenota_finestra):
        try:
            conn = connessione_db()
            c = conn.cursor()

            data_scelta = giorno_entry.get()
            orario_scelto = orario_combobox.get()
            sala_scelta = sala_var.get()

            orario_attuale = datetime.now().strftime("%H:%M")
            oggi = datetime.now().strftime("%Y-%m-%d")  # Ottieni la data di oggi

            # Verifica se il giorno selezionato è oggi
            if data_scelta == oggi:
                if orario_scelto < orario_attuale:
                    messagebox.showerror("Errore", "Non si può scegliere un orario passato per il giorno di oggi.", parent=self)
                    conn.close()
                    return

            # Controllo se la sala è stata selezionata
            if not sala_scelta:
                messagebox.showerror("Errore", "Seleziona una sala prima di confermare la prenotazione.", parent=prenota_finestra)
                return

            if datetime.strptime(data_scelta, "%Y-%m-%d").date() < datetime.now().date():
                messagebox.showerror("Errore", "Non puoi scegliere una data antecedente a oggi.", parent=prenota_finestra)
                return

            c.execute("SELECT * FROM prenotazioni WHERE sala = %s AND data = %s AND orario = %s", (sala_scelta, data_scelta, orario_scelto))
            if c.fetchone():
                messagebox.showerror("Sala occupata", "La sala selezionata è già occupata in questo orario.", parent=prenota_finestra)
                return

            c.execute("SELECT * FROM prenotazioni WHERE paziente_id = %s AND data = %s AND orario = %s", (self.paziente_id, data_scelta, orario_scelto))
            if c.fetchone():
                messagebox.showerror("Errore", "Hai già una prenotazione in questo orario.", parent=prenota_finestra)
                return

            if messagebox.askyesno("Conferma Prenotazione", f"Vuoi confermare la prenotazione per il {data_scelta} alle {orario_scelto} in {sala_scelta}?", parent=prenota_finestra):
                c.execute("INSERT INTO prenotazioni (paziente_id, data, infermiere_id, sala, orario, tipo_visita) VALUES (%s, %s, %s, %s, %s, %s)", (self.paziente_id, data_scelta, self.infermiere_id, sala_scelta, orario_scelto, "Infermieristica"))
                conn.commit()
                messagebox.showinfo("Successo", "Prenotazione confermata con successo!", parent=prenota_finestra)
                prenota_finestra.destroy()
            else:
                messagebox.showinfo("Annullato", "Prenotazione annullata.", parent=prenota_finestra)
        finally:
            conn.close()

    def aggiorna_comunicazioni_paziente(self):
        oggi = datetime.today().date()
        conn = connessione_db()
        c = conn.cursor()

        # Comunicazioni interne 
        c.execute("""
            SELECT 
                e.data, 
                CONCAT(m.Nome, ' ', m.Cognome) AS medico,
                CONCAT(i.Nome, ' ', i.Cognome) AS infermiere
            FROM esiti e
            LEFT JOIN medici m ON e.medico_id = m.id
            LEFT JOIN infermieri i ON e.infermiere_id = i.id
            WHERE e.paziente_id = %s AND e.data < %s
            ORDER BY e.data DESC
        """, (self.paziente, oggi))
        mie_comunicazioni = c.fetchall()

        # Recuperare le prenotazioni future del paziente
        c.execute("""
            SELECT p.data, p.orario, p.sala, p.tipo_visita,
                m.Nome AS NomeMedico, m.Cognome AS CognomeMedico,
                i.Nome AS NomeInfermiere, i.Cognome AS CognomeInfermiere
            FROM prenotazioni p
            JOIN pazienti pa ON p.paziente_id = pa.id
            LEFT JOIN medici m ON p.medico_id = m.id
            LEFT JOIN infermieri i ON p.infermiere_id = i.id
            WHERE p.paziente_id = %s AND p.data >= %s
            ORDER BY p.data, p.orario
        """, (self.paziente, oggi))
        tutte_prenotazioni_future = c.fetchall()
        
        c.close()

        mie_comunicazioni_str = "\n".join(
            [
                f"- È stato caricato l'esito del giorno {data.strftime('%Y-%m-%d')}\n" +
                (f"Medico: {medico}" if medico else f"Infermiere: {infermiere}")
                for data, medico, infermiere in mie_comunicazioni
            ]
        ) if mie_comunicazioni else "Nessuna comunicazione registrata."


        tutte_prenotazioni_future_str = "\n".join(
            [f"- {data.strftime('%Y-%m-%d')} alle {orario} in {sala} {tipo_visita}" +
            (f"\nMedico: {nome_medico} {cognome_medico}" if nome_medico else f" \nInfermiere: {nome_infermiere} {cognome_infermiere}")
            for data, orario, sala, tipo_visita, nome_medico, cognome_medico, nome_infermiere, cognome_infermiere in tutte_prenotazioni_future]
        ) if tutte_prenotazioni_future else "Nessuna prenotazione."


        # Aggiornare il contenuto delle Textbox
        self.textbox_mie_comunicazioni.configure(state="normal")
        self.textbox_mie_comunicazioni.delete("1.0", "end")
        self.textbox_mie_comunicazioni.insert("1.0", mie_comunicazioni_str)
        self.textbox_mie_comunicazioni.configure(state="disabled")

        self.textbox_tutte_comunicazioni.configure(state="normal")
        self.textbox_tutte_comunicazioni.delete("1.0", "end")
        self.textbox_tutte_comunicazioni.insert("1.0", tutte_prenotazioni_future_str)
        self.textbox_tutte_comunicazioni.configure(state="disabled")

    def logout(self):
        # Mostra una finestra di dialogo di conferma
        risposta = messagebox.askyesno("Conferma Logout", "Sei sicuro di voler uscire?", parent = self)
        
        if risposta:  # Se l'utente conferma
            # Chiudi la connessione al database se è aperta
            if self.conn is not None and self.conn.is_connected():
                self.conn.close()
            
            # Chiudi la finestra corrente
            self.destroy()

    def clear_frame(self):
        for widget in self.winfo_children():
            widget.destroy()

class ModificaDatiPazienteDialog(ctk.CTkToplevel):
    def __init__(self, parent, paziente_id):
        super().__init__(parent)
        self.paziente_id = paziente_id
        self.title("Modifica Dati Paziente")
        self.geometry("600x400")

        # Imposta la finestra come modale rispetto alla finestra principale
        self.transient(parent)  # Fa sì che la finestra sia sempre sopra la finestra principale

        self.conn = connessione_db()
        self.c = self.conn.cursor()
        self.c.execute("""
            SELECT p.Nome, p.Cognome, p.CS, p.Medico_Curante, p.Nascita, p.Luogo_Nascita, p.Email, p.Password, 
                   CONCAT(m.Nome, ' ', m.Cognome) AS medico_curante_nome
            FROM pazienti p
            LEFT JOIN medici m ON p.Medico_Curante = m.ID
            WHERE p.ID = %s
        """, (self.paziente_id,))
        self.dati = self.c.fetchone()

        # Recupera la lista dei medici per l'OptionMenu
        self.c.execute("SELECT id, CONCAT(Nome, ' ', Cognome) AS nome_completo FROM medici")
        self.lista_medici = self.c.fetchall()  # Restituisce una lista di tuple (id, nome_completo)
        
        self.entries = {}
        self.crea_form()

    def crea_form(self):
        fields = [
            ("Nome", self.dati[0]), 
            ("Cognome", self.dati[1]), 
            ("CS", self.dati[2]), 
            ("Medico Curante", self.dati[8]),
            ("Nascita", self.dati[4]), 
            ("Luogo Nascita", self.dati[5]),  
            ("Email", self.dati[6]), 
            ("Password", self.dati[7])
        ]
        
        # Variabile per tenere traccia della riga corrente
        row = 0
        
        for i, (field, value) in enumerate(fields):
            col = i % 2  # Colonna 0 o 1
            
            # Creazione e posizionamento della label
            label = ctk.CTkLabel(self, text=field)
            label.grid(row=row, column=col * 2, padx=10, pady=5, sticky="w")
        

            if field == "Medico Curante":

                medico_attuale = self.dati[8]  # Nome del medico curante attuale
                # Crea l'OptionMenu per il medico curante
                self.medico_var = ctk.StringVar(value=medico_attuale)
                medico_opzioni = [medico[1] for medico in self.lista_medici]
                
                medico_option_menu = ctk.CTkOptionMenu(self, values=medico_opzioni, variable=self.medico_var)
                medico_option_menu.grid(row=row, column=col * 2 + 1, padx=10, pady=5, sticky="w")
                self.entries[field] = medico_option_menu


            else:
                entry = ctk.CTkEntry(self)
                entry.insert(0, value)
                entry.grid(row=row, column=col * 2 + 1, padx=10, pady=5, sticky="w")
                self.entries[field] = entry
        
        
            if col == 1:  # Passa alla riga successiva dopo aver riempito entrambe le colonne
                row += 1
        
        ctk.CTkButton(self, text="Salva", command=self.salva_modifiche).grid(row=row+1, column=0, columnspan=4, pady=20)

    def salva_modifiche(self):
        nuovo_nome = self.entries["Nome"].get()
        nuovo_cognome = self.entries["Cognome"].get()
        nuovo_cs = self.entries["CS"].get()
        nuovo_medico_curante = self.entries["Medico Curante"].get()
        nuova_nascita = self.entries["Nascita"].get()
        nuovo_luogo_nascita = self.entries["Luogo Nascita"].get()
        #nuova_responsabilita = self.entries["Responsabilita"].get()
        nuova_email = self.entries["Email"].get()
        nuova_password = self.entries["Password"].get()

        try:
            self.c.execute("""
                UPDATE pazienti
                SET Nome = %s, Cognome = %s, CS = %s, Medico_Curante = %s, Nascita = %s, Luogo_Nascita = %s,  Email = %s, Password = %s
                WHERE ID = %s
            """, (nuovo_nome, nuovo_cognome, nuovo_cs, nuovo_medico_curante, nuova_nascita, nuovo_luogo_nascita, nuova_email, nuova_password, self.paziente_id))
            self.conn.commit()
            messagebox.showinfo("Successo", "Modifiche salvate con successo.", parent = self)
            self.destroy()
        except mysql.connector.Error as err:
            messagebox.showerror("Errore", f"Errore durante l'aggiornamento: {err}", parent = self)
        finally:
            self.conn.close()
## fine pazeinte

#  calendario
class Calendario:
    def __init__(self):
        self.conn = connessione_db()
        self.sala_rotations = deque()  # Coda circolare per mantenere la rotazione delle assegnazioni

    def retrieve_nurses(self):
        cursor = self.conn.cursor()
        cursor.execute("SELECT id FROM infermieri ORDER BY id")
        nurses = cursor.fetchall()
        cursor.close()
        # Restituisce una lista con solo gli ID degli infermieri
        # 'nurse[0]' accede all'ID di ciascun infermiere (poiché 'fetchall()' restituisce tuple)
        return [nurse[0] for nurse in nurses]
    
    def initialize_rotation(self, nurses):
        # Mescola casualmente la lista degli infermieri passata come argomento
        random.shuffle(nurses)
        # Aggiunge (estende) la lista mescolata degli infermieri alla lista 'sala_rotations'
        # 'self.sala_rotations' è una lista che probabilmente tiene traccia delle rotazioni in sala degli infermieri
        self.sala_rotations.extend(nurses)

    def generate_shifts(self):
        nurses = self.retrieve_nurses()
        if not nurses:
            messagebox.showerror("Errore", "Nessun infermiere trovato nel database.", parent=self)
            return
        
        if len(self.sala_rotations) < len(nurses):
            self.initialize_rotation(nurses)
        
        try:
            cursor = self.conn.cursor()
            today = datetime.now().date()
            
            # Trova l'ultimo giorno del mese
            # Utilizza la data attuale 'today', la imposta temporaneamente al giorno 28 e poi aggiunge 4 giorni.
            # Questo garantisce di arrivare sempre in un giorno del mese successivo, indipendentemente dal numero di giorni nel mese (funziona anche per i mesi con meno di 31 giorni).
            next_month = today.replace(day=28) + timedelta(days=4)  # Questo garantirà di essere sempre nel mese successivo
            # Calcola l'ultimo giorno del mese togliendo il numero di giorni corrispondenti all'attuale giorno di 'next_month'.
            last_day_of_month = next_month - timedelta(days=next_month.day)
           

            # Trova l'ultimo sabato del mese
            # Calcola l'ultimo sabato del mese partendo dall'ultimo giorno del mese e sottraendo i giorni necessari per arrivare all'ultimo sabato.
            last_saturday = last_day_of_month - timedelta(days=(last_day_of_month.weekday() + 1) % 7 + 1)
            
            # Controlla se la data odierna è uguale o successiva all'ultimo sabato
            if today >= last_saturday:
                # Se la data odierna è dopo l'ultimo sabato del mese, prepara i turni per il mese successivo.
                start_date = last_day_of_month + timedelta(days=1) # Inizio dal primo giorno del mese successivo
                end_date = start_date.replace(day=28) + timedelta(days=4) # Va al mese successivo
                end_date = end_date - timedelta(days=end_date.day)  # Ultimo giorno del mese successivo
            else:
                # Se la data odierna è prima dell'ultimo sabato del mese, i turni vengono creati per il mese corrente
                start_date = today
                end_date = last_saturday

            # funziona, basta cambiare la data al pc e vengono generati i turni
            
            # Questo ciclo serve per creare i turni nel periodo compreso tra 'start_date' ed 'end_date'
            while start_date <= end_date:
                if start_date.weekday() < 5:   # Controlla se il giorno è dal lunedì al venerdì
                    cursor.execute("SELECT COUNT(*) FROM calendario WHERE data = %s", (start_date,))
                    # Controlla se ci sono già turni inseriti per quella data nel calendario.
                    if cursor.fetchone()[0] == 0:
                        # Se non ci sono turni per quella data, vengono assegnati i turni
                        sala_prelievi = self.sala_rotations[0] # Seleziona l'infermiere per la sala prelievi
                        sala_medicazioni = self.sala_rotations[1] # Seleziona l'infermiere per la sala medicazioni

                        # Inserisce i turni nel database
                        cursor.execute("""
                            INSERT INTO calendario (data, sala_prelievi, sala_medicazioni)
                            VALUES (%s, %s, %s)
                        """, (start_date, sala_prelievi, sala_medicazioni))
                        
                        # Ruota la lista degli infermieri per garantire che cambino turno nel giorno successivo.
                        self.sala_rotations.rotate(-1)
                
                # Passa al giorno successivo
                start_date += timedelta(days=1)

            self.conn.commit()
        except Exception as e:
            messagebox.showerror("Errore", "Errore durante l'inserimento dei turni: {str(e)}", parent=self)
        finally:
            cursor.close()

    def close(self):
        self.conn.close()
    
if __name__ == "__main__":

    calendario = Calendario()
    calendario.generate_shifts()  # Genera i turni

    app = Application()
    app.mainloop()
```
